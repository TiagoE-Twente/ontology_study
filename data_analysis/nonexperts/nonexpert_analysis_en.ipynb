```python
import pandas as pd
from collections import Counter
from scipy.stats import chi2_contingency
import numpy as np

# Load your Excel file (replace the path with your actual file)
file_path = '/content/sample_data/nonexperts_EN_clean.xlsx'
df = pd.read_excel(file_path)

# Example correct answers dictionary
correct_answers = {
    "Brocolli_img": [
        "Brocolli",
        "Broccoli",
        "Broccoli",
        "Brocoli",
        "Broccoli",
        "Broccoli",
        "Broccoli",
        "Brocoli",
        "Broccoli",
        "Broccoli",
        "broccoli",
        "broccoli",
        "Br√≥coli",
        "Broccoli",
        "broccoli",
        "Broccoli",
        "Broccoli",
        "Brocoli",
        "Broccoli",
        "Broccoli",
        "broccoli",
    ],
    "Carrot_img": [

        "Carrots",
        "Carrot",
        "carrot"
        "Carrots",
        "Carrots",
        "Carrots",
        "Carrots",
        "carrots",
        "carrots",
        "Carrots",
        "carrots",
        "Carrots",
        "Carrots",
        "Carrots",

    ],
    "Cauliflower_img": [
        "Colliflower",
        "Couliflower",
        "Cauliflower",
        "Cauliflowers",
        "Cauliflower",
        "Cauliflower",
        "Cauliflower",
        "Cauliflower",
        "Cauliflower",
        "cauliflower",
        "cauliflower",
        "Cauliflower",
        "cauliflower",
        "Califlower",
        "Cauliflower",
        "Cauliflower",
        "Cauliflower",
        "Cauliflower",
    ],
    "Chicken_Breast_img": [

        "Chicken Breast",


        "Chicken breast",

        "Chicken breast",
        "Chicken breast",
        "Chicken breast",
        "Chicken breast",
    ],
    "Apple_img": [
        "Apple",
        "Apple",
        "Apple",
        "apples",
        "Apples"


        "Apple",


        "apple",


        "Apple",
        "Apple",



    ],
    "Ham_slice_img": [
        "Ham",

        "Ham",
        "Ham",
        "Ham",

        "Ham",
        "hams",
        "Ham",
        "Ham",
        "Ham",

        "ham",
        "Ham",

        "Ham",
        "Ham",
        "Ham",

    ],
    "McIntosh_Apple_img": [
        "Apple",
        "Apple",
        "Apple",

        "Apple",
        "apples",

        "Apple",

        "Apple",
        "Apple",
        "apple",
        "apple",

        "Apple",
        "Apple",

        "Apple",

        "Apple",
    ],
    "moon_drop_grape_img": [

        "Grape",


        "Grape",


        "Grapes",


        "Grapes",




        "Grapes",

    ],
    "pork_sausage_img": [
        "Sausage",
        "Sausage",
        "Sausages",
        "sausage",

        "Sausage",
        "Sausage",



        "Sausage",
        "Sausage",
        "sausages",
        "sausages",

        "sausages",
        "Sausage",
        "Sausage",

        "Sausages",

        "Sausages",
    ],
    "red_globe_grape_img": [
        "Grape",
        "Grapes",
        "Grapes",
        "Grapes",

        "Grape",
        "Grapes",


        "Grapes",
        "Grapes",

        "grapes",
        "Grapes",

        "Grapes",
        "Grapes",

        "Grapes",


    ],
    "ribeye_steak_img": [
        "Steak",

        "Steak",

        "Steak",

        "Steak",
        "Steak",

        "Steak",

        "Steak",


        "Steak",

    ],
    "salmon_fillet_img": [
        "Salmon",
        "Salmon",
        "Salmon",
        "Salmon",

        "Salmon",
        "Salmon",

        "Salmon",
        "Salmon",

        "salmon",
        "salmon",

        "Salmon",

        "Salmon",

    ],
    "sweet_potato_img": [

        "Sweet potato",
        "Sweet Potato",
        "Sweet potato",
        "Sweet potatoe",

        "Sweet potatoe",

        "Sweet potato",
        "Sweet potatoes",

        "sweet potatoes",
        "Sweet potatoes",
        "Seet potatoes",

        "Sweet potato",

        "Sweet Potato",
        "Sweet potato",
        "Sweet potatoe",


    ],
    "t_bone_steak_img": [
        "T bone steak",


        "Mmmmmmm t bone steak",

        "T-bone steak",

        "T-bone steak",
        "T bone steak",




        "Raw T Bone Steak",


    ],
    "tuna_steak_img": [


        "Tuna",

        "Tuna",
        "Tuna?",


        "Tuna",
        "Tuna",


        "Tuna",

        "Tuna",
        "Tuna",
        "Tuna",


    ],
    "turkey_breast_img": [

        "Cooked chicken",

        "Cooked chicken breast",
        "cooked chicken",

        "Cooked chicken",

    ],
    "bass_trombone_img": [
        "Trombone",
        "Trombone",
        "Trombone",

        "Trombone",


        "Trombon",
        "Trombone",

        "trombone",



        "Trombone",

        "Trombone",

        "Trombone",
    ],
    "bell_img": [
        "Bell",
        "Bell",
        "Bell",
        "Bell",

        "Bell",

        "Bell",
        "Bell",
        "Bell",
        "bell",
        "bell",
        "Bell",

        "bell",
        "Bell",

        "Bell",
        "Bell",

    ],
    "bongo_drums_img": [

        "Bongos",

        "Bongos",

        "Bongos",

        "Bongos",

        "bongos",

        "Bongos",


        "Bongos",


        "Bongos",


    ],
    "castanets_img": [
        "Wood",
        "Castanet",

        "Castanets",

        "Castanets???",
        "Castanets",
        "Casta√±uelas",



        "castanets",
        "Casta√±uelas",
        "Casta√±uelas",


        "Castanets",
        "Castanets",

    ],
    "clarinet_img": [

        "Clarinet",
        "Clarinet",
        "Clarinet",

        "Clarinet",
        "Clarinet",


        "Clarinet",

        "clarinet",
        "clarinet",
        "Clarinete",


        "Clarinet",
        "Clarinet",
        "Clarinet",

    ],
    "cornet_img": [
        "Ttumprt",
        "Trumpet",
        "Trumpet",

        "Trumpet",

        "Trumpet",
        "Trumpet",
        "trumpet",
        "trumpet",
        "Trumpet",
        "Teumpet",
        "trumpet",
        "Trumpet",
        "Trumpet",

        "Trumpet",

    ],
    "fife_img": [

        "Flute",
        "Flute",
        "Flute",

        "Flute",
        "Flute",
        "Flute",



        "Flute",
        "flute",
        "flute",


        "flute",
        "Flute",


        "Flute",

        "Flute",
    ],
    "piccolo_img": [

        "Flute",


        "Flute",

        "flute",
        "Flute",

        "flute",

        "Flute",


    ],
    "roto_drums_img": [

        "Drums",
        "Drums",
        "Drums",
        "Drums",
        "Drum",
        "DrumS",
        "Drums",

        "Drums",
        "Drums",


        "drums",


        "drums",


    ],
    "saxophone_img": [
        "Saxophone",
        "Sax",
        "Saxophone",
        "Saxofone",

        "Saxophone",


        "Saxo",
        "Saxiphone",
        "Saxophone",
        "saxophone",
        "saxophone",
        "Saxophone",

        "saxophone",
        "Saxophone",
        "Saxophone",
        "Saxophone",
        "Saxophone",
        "Saxophone",

        "Saxophone",
    ],
    "slit_drum_img": [

        "Xylophone",


        "Xylophone",

        "Xylophone",
        "Xilophone",


        "Xylophone",
        "Xylophone",
        "xylophone",
        "xylophone",
        "Xylophone",
        "Xilophone",

    ],
    "tambourine_img": [

        "Tambourine",
        "Tambourine",
        "Tamborine",

        "Tambourine",
        "Tambourine",
        "Tambourine fun",

        "Tamborine",

        "tambourine",
        "tambourine",

        "tambourine",

        "Tamborine",
        "Tambourine",

        "tamborine",

    ],
    "tenor_trombone_img": [
        "Trombone",
        "Trombone",
        "Trombone",
        "Trombone",

        "Trombone",
        "Trombone music",

        "Trombone",


        "trombone",
        "Tromb√≥n",

        "Trombone",

        "Trombone",

        "Trombone instrument",

        "Trompeta",

    ],
    "timpani_img": [
        "Drum",

        "Drum",

        "Drum",
        "Drum",

        "Drum",

        "Drum",
        "Drum",
        "Drum",
        "drum",
        "drum",
        "Drum",

        "drum",

        "Drum",

        "Drums",
    ],
    "triangle_img": [
        "Triangle",
        "Triangle",
        "Triangle",
        "Triangle (instrument)",

        "Triangle",
        "Triangle",

        "Triangle",
        "Triangle",
        "Triangle",
        "triangle",
        "triangle",
        "Triangle",

        "triangle",

        "Triangle",
        "Triangle",
        "Triangle",

    ],
    "trumpet_img": [
        "Trumpet",
        "Trumpet",
        "Trumpet",
        "Trumpet",

        "Trumpet",
        "Trumpet",



        "Trumpet",
        "Trumpet",
        "trumpet",

        "Trumpet",
        "Trumpet",
        "trumpet",
        "Trumpet",





    ],
    "3D_printer_img": [
        "3d printer",
        "3d printer",
        "3D Printer",
        "3D printer",
        "3d printer",
        "3d printer",

        "3d printer",

        "3D printer",
        "3D printer",


        "3D printer",
        "3d printer",

        "3d printen",
        "3d printer",


        "3D printer",



    ],
    "alienware_desktop_img": [
        "speaker",
        "Speaker",
        "Speaker",

        "speakers",
        "Speaker",
        "Speaker",
        "speaker",
        "speaker",
        "Speaker",
    ],
    "alienware_laptop_img": [

        "Laptop",
        "Laptop",


        "Laptop",

        "Laptop",
        "Laptop",
        "Laptop",
        "Laptop",

        "laptop",

        "Laptop",

        "laptop",

        "Laptop",
        "Laptop",
        "Lap top",
        "A laptop",
    ],
    "CRT_monitor_img": [

        "Computer monitor",

        "Old monitor",

        "monitor",


        "monitor",

        "Monitor",
        "Monitor",

        "monitor",

        "Monotor",
    ],
    "iMac_img": [

        "Apple computer",


        "Mac computer üñ•Ô∏è",




        "Computer",

        "Computer",


        "computer",
        "Computer",
        "Monitor",

        "Apple computer",
    ],
    "LCD_monitor_img": [

        "Monitor",

        "Monitor",
        "Monitor",
        "monitor",


        "Monitor",



        "Monitor",

        "monitor",
        "Monitor",
        "Monitor",
        "Monitor",
        "monitor screen",
        "monitor",

    ],
    "mac_pro_img": [


        "mac pro",

        "Mac studio",

    ],
    "macbook_air_img": [

        "Laptop",
        "Laptop",




        "Laptop",
        "Laptop",
        "Laptop",
        "Laptop",

        "laptop",
        "laptop",
        "Laptop",
        "Laptop",
        "Laptop",
        "laptop",
        "laptop",


    ],
    "macbook_pro_img": [

        "Laptop",
        "Laptop",
        "Laptop",
        "Laptop",

        "Laptop",

        "Laptop",

        "Laptop",
        "Laptop",
        "Laptop",
        "laptop",
        "laptop",
        "Laptop",
        "Laptop",
        "Laptop",

        "laptop",

        "LapTop",
    ],
    "wired_keyboard_img": [
        "keyboard",
        "Keyboard",

        "Keyboard",

        "keyboard",
        "Keyboard",
        "keyboard",
        "Keyboard",
        "Keyboard",

        "Keyboard",

        "keyboard",
        "keyboard",
        "Keyboard",
        "Keyboard",
        "Keyboard",
        "keyboard",

        "Keyboard",
        "Keyboard",
    ],
    "mouse_img": [
        "mouse",
        "Mouse",

        "Mouse",

        "mouse",

        "mouse",
        "Mouse",
        "Mouse",
        "Mouse",


        "mouse",
        "mouse",
        "Mouse",
        "Mouse",
        "Mouse",
        "mouse",

        "Mouse",

        "Mause",
    ],
    "wireless_keyboard_img": [

        "Keyboard",

        "Keyboard",


        "Keyboard",
        "Keyboard",
        "Keyboard",
        "Keyboard",

        "keyboard",
        "keyboard",
        "Keyboard",
        "Keyboard",
        "Keyboard",

        "keyboard",


    ],
    "wireless_mouse_img": [
        "mouse",
        "Mouse",

        "Mouse",
        "Mouse",
        "mouse",

        "mouse",

        "Mouse",
        "Mouse",
        "Mouse",

        "mouse",
        "mouse",
        "Mouse",
        "Mouse",
        "Mouse",
        "mouse",


        "Mouse",
    ],
    "XPS_desktop_img": [

        "Personal Computer",

        "Pc old",

        "Pc tower",


        "PC",

        "Pc",

        "PC with Intel icore i7",
        "Pc",
    ],
    "XPS_laptop_img": [
        "laptop",
        "Laptop",
        "Laptop",
        "Laptop",

        "laptop",

        "laptop",
        "Laptop",

        "Laptop",

        "Laptop",
        "laptop",
        "laptop",
        "Laptop",
        "Laptop",
        "Laptop",

        "laptop",

        "Lap top",

        "A Laptop",
    ],
    "chainsaw_img": [

        "Chainsaw",
        "Chainsaw",
        "Chainsaw",
        "Chainsaw",
        "chainsaw",
        "Chainsaw",


        "Chainsaw",



        "chainsaw",
        "Chainsaw",
        "Chainsaw",
        "Chain saw",
        "chainsaw",

        "Chain saw",
        "Chainsaw",


        "Chainsaw",
    ],
    "chisel_img": [

        "Chisel",
        "Chisel",


        "chisel?",






        "Chisel",

        "Chisel",

        "groove chisel",

    ],
    "drain_shovel_img": [

        "Shovel",
        "Shovel",
        "Shovel",
        "Shovel",


        "shovel",

        "Shovel",

        "Shovel",
        "Shovel",

        "shovel",
        "Shovel",
        "Shovel",

        "shovel",


        "Shovel",
    ],
    "electric_pruner_img": [




        "Cutters",

        "Cutter",

        "cutter",


        "Cutter",

    ],
    "flat_brush_img": [
        "Paintbrush",

        "Paintbrush",
        "Paintbrush",
        "Paintbrush",
        "paint brush",

        "paintbrush",
        "Paintbrush",
        "Paint brush",

        "paintbrush",

        "Paint brush",

        "paint brush",


    ],
    "gardening_shear_img": [
        "clippers",

        "Clipper",
        "clippers",
        "Clippers",

    ],
    "glass_palette_img": [



        "Watercolors",

        "Water colors",
        "Watercolors",

        "water color pallette",
        "Water paint",
        "Watercolor pallet",

    ],
    "lopper_img": [

        "Cutter",

        "Branch cutter",



        "cutter",
        "cutter",



        "branch cutter",


    ],
    "mallet_img": [
        "mallet",

        "Mallet",

        "Mallet",
        "meat mallet",

        "mallet",
        "Mallet (meat tenderizer",


        "Meat mallet",


        "Mallet",





    ],
    "potting_trowel_img": [
        "shovel",
        "Mini shovel",

        "Small shovel",
        "Shovel for crops",
        "small shovel",

        "shovel",

        "Shovel",





        "Luh shovel",
        "Shovel",
        "small shovel",


        "Shovel",

    ],
    "rake_img": [
        "rake",
        "Rake",
        "Rake",
        "Rake",
        "Rake",
        "rakes",
        "Rake",
        "rake",
        "Rake",
        "Rake",
        "Rake",
        "Rake",
        "Rake",




        "Rake",

        "rake",
        "Rake",


    ],
    "rockery_trowel_img": [

        "Tool",

        "Carving tool",


        "Tool",





    ],
    "round_brush_img": [
        "paintbrush",

        "Paintbrush",
        "Paint brush",
        "Paintbrush",
        "paint brush",

        "paintbrush",
        "Paintbrush",
        "Paint brush",

        "paintbrush",

        "Paint brush",
        "Paint brush",
        "paint brush",
        "paint brush",

        "Paint brush",
    ],
    "rounded_point_shovel_img": [
        "shovel",
        "Shovel",
        "Shovel",
        "Shovel",
        "Shovel",
        "shovel",
        "Shovel",
        "shovel",

        "Shovel",

        "Shovel",
        "Shovel",

        "shovel",
        "Shovel",
        "Shovel",

        "shovel long",
        "shovel",

    ],
    "wire_img": [
        "wire",

        "Metal wire",
        "Wire",

        "wire",
        "Wire",
        "wire",
        "Wire",

        "Wire",
        "Wire",
        "Wire",
        "wire",

        "Wire",
        "Wire",
        "Wire",

        "wire",

    ]

    ,
    "wood_palette_img": [
        "paint palette",

        "Painter palette",

        "Paint pallet? üé® stuff to put paint on",
        "paint palette",


        "Paint palate",
        "Paint pallet",
        "Painter‚Äôs palette",

        "Palette",



        "Palette",

        "palette board",

    ],
    "apple_tree_img": [
        "Apple tree",

        "Apple tree",



        "Apple tree",
        "Apple tree",

        "Apple trees",
        "apple tree",

        "Apple tree",
        "Apple orchard",

        "Apple tree",
        "apple tree?",

        "apple tree",
        "apple tree",
        "apple trees",

        "Apple tree",
    ],
    "bulldog_img": [

        "bulldog",

        "bulldog",
        "Bulldog",

        "Bull Dog",

        "Bulldog",


        "Bulldog",
        "Bulldog",
        "bulldog",
        "Bulldog",

        "Bull dog",

        "bull dog",


        "Bulldog",

    ],
    "cedar_img": [
        "Tree",
        "tree",
        "Tree",
        "Big tree",

        "Tree",
        "Tree",

        "A middle age tree",

        "tree",

        "Tall tree",
        "Tall tree",
        "tree",
        "Tree",
        "tree",

        "tree",
        "Tall tree",

        "tree",
        "Tree",
    ],
    "dandelion_img": [

        "yellow flower",


        "Yellow flower",

        "Yellow flower",




        "Yellow flower",

        "Picture of an exotic, yellow flower",
        "yellow flower",
        "yellow flower",
        "Yellow flower",
    ],
    "Eagle_img": [
        "Eagle(Freedom)",
        "eagle",

        "eagle",
        "Eagle",
        "Eagle/bird",


        "Eagle",
        "Eigle",

        "eagle",
        "Eagle",
        "eagle",
        "An eagle",
        "eagle",

        "eagle",
        "eagle",

        "Eagle",
    ],
    "Edelweiss_img": [
        "Flower",

        "Flower",

        "Flower",
        "Flower",


        "Flower",



        "Flower",

        "flower",


    ],
    "german_shepherd_img": [
        "German Shepard",
        "german shephard",
        "Dog (german shepherd)",

        "german sheppard",

        "German shepherd",
        "Germ√°n shepherd",
        "German shepherd",
        "German Shepard",
        "German sheperd",
        "German Shepherd",
        "German shepherd",
        "german shepherd",

        "shephard",
        "German shepherd",

        "German Shepard - dog",

        "German shepherd",
    ],
    "orange_tree_img": [
        "Orange tree",
        "orange tree",
        "Orange tree",
        "An Orange tree",
        "orange tree",
        "Orange tree",
        "Orange tree",
        "Orange tree",

        "Orange yree",
        "oranges tree",
        "Orange tree",
        "Orange tree",
        "Orange tree",
        "orange tree",
        "Orange tree",
        "orange tree",
        "Orange tree",
        "orange tree",
        "Orange Tree",
        "orange trees",
        "orange tree",
        "Orange tree",
    ],
    "Ostrich_img": [
        "Ostrich",
        "ostrich",
        "Ostrich",

        "ostrich",
        "Ostrich",

        "Ostrich",

        "Ostrich",
        "ostrich",

        "Ostrich",
        "Ostrich",
        "ostrich",
        "Ostrich",
        "ostrich",
        "Ostrich",
        "ostrich",
        "Ostrich",

        "ostrich",
        "Ostrich",
        "Ostrich",
        "an ostrich",
    ],
    "owl_img": [

        "owl",
        "Owl",

        "owl",
        "Owl",
        "Bird/owl",
        "Owl",

        "Owl",
        "owl",
        "Owl",
        "Owl",

        "owl",
        "Owl",
        "owl",
        "Owl",
        "owl",

        "owl",
        "owl",
        "Owl",
        "Owl",
    ],
    "penguin_img": [
        "Skipper(Penguin)",
        "penguin",
        "Penguin",

        "penguin",
        "Penguin",
        "Penguin",
        "Penguin",
        "Penguin",
        "Penguin",
        "Penguin",

        "Penguin",

        "penguin",
        "Penguin",
        "penguin",
        "Penguin",
        "penguin",

        "penguin",
        "pinguin",
        "Penguin",
    ],
    "persian_img": [

        "cat",
        "Cat",

        "cat",
        "Cat",
        "Cat",
        "Cat",

        "Cat",


        "cat",
        "Cat",
        "cat",
        "Cat",
        "cat",


        "fluffy cat",
        "Cat",
        "Ugly cat",
    ],
    "Pine_img": [
        "tree",

        "Tree",

        "tree (known for height)",
        "Tree",
        "Tree",
        "Tree",

        "tree",

        "Tree",
        "Tree",
        "tree",
        "Tree",
        "tree",

        "Tree",

        "tree",
        "Tree",
    ],
    "rose_img": [

        "Rose",

        "wet rose",
        "Red rose",
        "Rose",
        "Red rose",

        "Red rose",


        "Rose",

        "rose",
        "Wet rose",
        "rose",
        "Red rose",
        "flower",

        "red rose",
        "red rose",
        "Rose",
        "wet rose",
        "Rose",
    ],
    "siamese_img": [
        "Cat",
        "cat",
        "Cat",

        "cat",

        "Cat",

        "Cat",


        "cat",
        "Cat",
        "cat",

        "cat",


        "cat",
        "Cat",
    ],
    "tulip_img": [
        "Flower",
        "flower",
        "Flower (tulip)",

        "flower",

        "Flower",

        "Flower",
        "Flower",


        "flower",

        "flower",


        "Flower",
        "Flower",

    ],
    "cerberus_img": [

        "Cerberus",


        "Cerberus",


        "Cerbero",

        "Cerberus",

        "cerberus",
        "Cerberus",


        "Cerberus",


    ],
    "cherubim_img": [
        "Angel",
        "angel portray",
        "Angel",

        "Cherub/angel",
        "Angel",

        "Sweet Angel",
        "Angel",
        "angel",
        "Angel kid",

        "angel",
        "Angel",

        "Angel",


        "baby angel",
        "Angel",
    ],
    "drow_img": [
        "Dark elf",


        "elf",

        "Archer elf",

        "elf",


        "elf",

    ],
    "forest_elf_img": [
        "Elf",

        "Wood Elf",


        "Elf warrior",
        "Elf",

        "Elf lady in the woods",
        "elf",

        "Elf in forest",



    ],
    "golem_img": [
        "Golem",

        "Golem",
        "Golem",


        "golem",

        "golem",


        "golem",

        "Golem",


    ],
    "harpy_img": [
        "Harpy",

        "Harpy",


        "Harpy",



        "harpy",

        "Harpy",
        "Harpie",
        "harpies",


        "harpie or siren",
    ],

    "hellhound_img": [

        "fire breathing dog",


        "Fire dog",

        "Fiery dog",

        "Dog with fire mouth",
        "Fire breathing dog",
        "Dog breathing fire",

        "Fire-breathing dog",
        "dog barking flames",



    ],
    "hydra_img": [

      "Hydra",

         "hydra",


    ],
    "mermaid_img": [
        "Mermaid",

        "Mermaid",

        "mermaid",
        "Mermaid",
        "Mermaid",
        "Mermaid",


        "mermaid",
        "Mermaid",
        "Mermaid",
        "Mermaid",

        "Mermaid",
        "mermaid",



        "mermaid",
        "mermaid",
        "Mermaid",
        "Mermaid",

    ],
    "pegasus_img": [
        "Pegasus",

        "Pegasus",

        "pegasus",
        "Pegasus",

        "Pegasus",


        "Pegaso",
        "Pegasus",
        "Pegasus",
        "pegasus",
        "Pegasus",

        "Pegassus",
        "pegasus",


        "Pegasus",

    ],
    "phoenix_img": [
        "Pheonix",
        "phoenix",
        "Phoenix",

        "phoeniz",
        "Phoenix",


        "Phoenix",
        "Fenix",

        "Phoenix",
        "Pheonix",



        "Phoenix image",
        "pheonix",

        "phoenix",

    ],
    "thunderbird_img": [
        "eagle",

        "Eagle (griffon like)",

        "Eagle",

        "Eagle",

        "Eagle",


        "Eagle",

        "eagle",
        "Eagle",


        "eagle",



        "Eagle",
    ],
    "triton_img": [
        "Poseidon",

        "Poseidon",

        "poseidon",

        "Poseidon",
        "Poseidon",

        "Poseidon",

        "Poseidon",

        "poseidon",


        "Poseidon",


        "poseidon",

    ],
    "troll_img": [
        "Troll",

        "Troll",


        "Troll",


        "troll",

        "troll",
        "Troll",
        "troll",


        "Troll",


    ],
    "unicorn_img": [
        "unicorn",
        "unicorn",
        "Unicorn",

        "unicorn",
        "Unicorn",
        "Unicorn",
        "Unicorn",
        "Unicorn",
        "Unicorn",
        "unicorn",
        "Unicorn",
        "Unicorn",

        "unicorn",

        "unicorn",

        "unicorn",

        "unicorn",
        "unicorn",
        "Unicorn",
    ],
    "wyvern_img": [
        "dragon",
        "dragon",

        "dragon",
        "Dragon",
        "Dragon",
        "Dragon",

        "Dragon",



        "dragon",

        "dragon",

        "dragon",


        "dragon",

    ],
}


# =========================
# Normalization
# =========================
# Clean the dataframe labels
df['Image'] = df['Image'].str.strip().str.lower()

# Clean the correct_answers dictionary
correct_answers_normalized = {
    k.lower().strip(): [ans.lower().strip() for ans in v]
    for k, v in correct_answers.items()
}

# =========================
# Agreement Rate Calculation with Debugging
# =========================
image_agreement = {}
missing_images = []
unmatched_responses = {}

for image in df['Image'].unique():
    if image not in correct_answers_normalized:
        missing_images.append(image)
        continue

    image_df = df[df['Image'] == image]
    responses = image_df['Response'].dropna().str.strip().str.lower()

    acceptable_answers = correct_answers_normalized[image]

    agree_count = 0
    disagree_count = 0
    unmatched_resp = []

    for resp in responses:
        if resp in acceptable_answers:
            agree_count += 1
        else:
            disagree_count += 1
            unmatched_resp.append(resp)

    total = agree_count + disagree_count
    agreement_rate = (agree_count / total) * 100 if total > 0 else 0

    image_agreement[image] = {
        'Agreement %': agreement_rate,
        'Agree Count': agree_count,
        'Disagree Count': disagree_count,
        'Total Responses': total
    }

    if unmatched_resp:
        unmatched_responses[image] = Counter(unmatched_resp)

# =========================
# Chi-Square Test
# =========================
# Prepare contingency table
contingency_data = []
labels = []

for image, data in image_agreement.items():
    if data['Total Responses'] >= 5:  # Ensure enough sample size
        contingency_data.append([data['Agree Count'], data['Disagree Count']])
        labels.append(image)

# Run chi-square if we have enough images
if len(contingency_data) >= 2:
    chi2, p, dof, expected = chi2_contingency(contingency_data)

    print("\nChi-Square Test Results")
    print(f"Chi2 Statistic: {chi2}")
    print(f"p-value: {p}")
    print(f"Degrees of Freedom: {dof}")
    print(f"Expected Frequencies:\n{expected}")

    # Interpret
    alpha = 0.05
    if p < alpha:
        print("\nConclusion: There is a statistically significant difference in agreement rates across images (reject null hypothesis).")
    else:
        print("\nConclusion: No significant difference in agreement rates across images (fail to reject null hypothesis).")
else:
    print("\nNot enough images with sufficient responses to perform chi-square test.")

# =========================
# Outputs
# =========================
print("\nAgreement Rates per Image:")
for image, stats in image_agreement.items():
    print(f"{image}: {stats['Agreement %']:.2f}% (n={stats['Total Responses']})")

if missing_images:
    print("\nMissing Images from Dictionary:")
    print(missing_images)

if unmatched_responses:
    print("\nUnmatched Responses by Image:")
    for image, counts in unmatched_responses.items():
        print(f"{image}: {dict(counts)}")
```
