```python
import pandas as pd
from collections import Counter
from scipy.stats import chi2_contingency
import numpy as np

# Load your Excel file (replace the path with your actual file)
file_path = '/content/sample_data/NONEXPERTS_SP_CLEAN (1).xlsx'
df = pd.read_excel(file_path)

# Example correct answers dictionary
correct_answers = {
    "apple_tree_img": [

        "Manzano",
        "manzano",

        "Manzano",

        "manzano",
        "Manzano",



        "Manzano",



        "Manzanos",
    ],
    "bulldog_img": [


        "Bulldog inglés",
        "Bull dog",

        "Bulldog",

        "Un bulldog",
        "bulldog",


        "Buldog",
        "Bulldog",


        "Bull dog",
        "bulldog",
        "Bulldog",

        "bull dog",

    ],
    "cedar_img": [
        "Arbol",

        "Árbol",
        "Arbol",


        "Arbol",
        "Árbol",
        "Un árbol",
        "arbol",
        "Arbol",


        "Arbol",

        "Un árbol",


        "Arbol",
        "Árbol",

        "UN ÁRBOL",

        "Arbol",
        "Arbol",
    ],
    "dandelion_img": [


        "Flor",

        "Flor",
        "Flor",

        "Flor",



        "Una flor",


        "Flor",
        "una flor",

        "Flor",

        "UNA FLOR",

    ],
    "Eagle_img": [
        "Aguila",
        "Aguila",
        "Aguila",
        "Águila",

        "Aguila",

        "Águila",

        "Aguila",

        "Aguila",
        "Águila",

        "Aguila",

        "Aguila",
        "Águila",
        "Águila",
        "UN ÁGUILA",

    ],
    "Edelweiss_img": [
        "Flor",

        "Flor",

        "flor",
        "Flor",
        "Flor",
        "Flor",

        "Flor",

        "Flor",
        "flor",
        "Flor",
        "Flor",

        "UNA FLOR",
        "flor",

        "Flor",
    ],
    "german_shepherd_img": [
        "Pastor aleman",
        "Pastor Aleman",
        "Pastor alemán.",
        "Pastor alemán",

        "Pastor aleman",

        "Pastor Alemán",

        "pastor aleman",


        "Pastor aleman",
        "Pastor Aleman",
        "Pastor alemán",

        "Pastor aleman",

        "Pastor alemán",

        "Pastor Alemán",

        "pastor aleman",

    ],
    "orange_tree_img": [

        "Naranjo",
        "Naranjo",
        "naranjo",

        "Naranjo",
        "Naranjo",
        "Un Naranjo",
        "naranjas",
        "Naranjo",


        "Naranjo",

        "Naranjo",

        "Naranjo",

        "naranjo",
        "Naranjos",
        "Un naranjo",
    ],
    "Ostrich_img": [
        "Avestruz",
        "Avestruz",
        "Avestruz",
        "Avestruz",
        "avestruz",
        "Avestruz",

        "Avestruz",
        "Una avestruz",
        "avestruz",
        "Avestruz",

        "Avestruz",


        "Un avestruz",
        "Avestruz",
        "un avestruz",
        "Avestruz",
        "Avestruz",
        "Avestruz",
        "UNA AVESTRUZ",


        "Avestruz",
    ],
    "owl_img": [
        "Buho",

        "Buho",
        "buho",
        "Buho",

        "Un búho",

        "Buho",

        "Búho",
        "un buho",

        "Búho",
        "buho",

        "Búho",
        "Buho",

        "buho",
        "Buho",
        "Buho",
        "Buho",
    ],
    "penguin_img": [
        "Pinguino",

        "Pinguino",
        "Pingüino",
        "pinguino",
        "Pingüino",

        "Pingüino",
        "Pingüino",

        "Pinguino",


        "Pinguino",
        "Pingüino",
        "pinguino",
        "Un pinguino",
        "Pinguino",
        "un pingüino",
        "Pingüino",
        "Pingüino",
        "Pinguino",
        "UN PINGÜINO",
        "pnguino",
        "Pingüino",
    ],
    "persian_img": [
        "Gato",

        "Gato",


        "Un gato",

        "Gato",


        "Gato",
        "Gato",


        "Gato",
        "gato",
        "Gato",
        "Gato",
        "Gato",

        "Gato",

    ],
    "Pine_img": [
        "Arbol",

        "Árbol",
        "Arbol",

        "Árbol",
        "Arbol",

        "Un árbol",
        "arbol",
        "Arbol",



        "Un arbol",
        "Árbol",
        "un arbol",

        "Árbol",

        "UN ÁRBOL",

    ],
    "rose_img": [
        "Rosa",

        "Rosa",

        "rosa",
        "Rosa",

        "Rosa",
        "Una rosa",
        "rosa",
        "Rosa",


        "Una rosa",


        "Rosa",

        "Rosa",

        "UNA ROSA",
        "rosa",

    ],
    "siamese_img": [
        "Gato",

        "Gato",

        "Gato",

        "Un gato",

        "Gato",

        "Gato",


        "un gato",

        "Gato",
        "Gato",

        "Gato",

    ],
    "tulip_img": [

        "Tulipan",

        "Tulipán",

        "Tulipán",

        "tulipan",
        "Tulipán",

        "Tulipan",


        "Tulipán",

        "Tulipan",

        "tulipan",

    ],
    "cerberus_img": [

        "Perro mitologico de tres cabezas cuidaba puerta del infierno  Cancerbero",

        "Cerbero",
        "Cerbero",
        "cerbero",

        "Cerbero",


    ],
    "cherubim_img": [
        "Angel",
        "Angel",


        "Angel",
        "Angel",
        "Ángel",
        "Un ángel",

        "Angelito",


        "Un ángel",

        "Un angel",
        "Angel",

        "Angel",

        "UN ANGELITO",

        "Angelito",
    ],
    "drow_img": [


        "Elfo",

        "Elfa",


        "Elfo Oscuro",


        "Elfa heroina",
        "Elfa",


        "Elfo",

    ],
    "forest_elf_img": [
        "Cazadora",

        "Cazadora",




        "Una cazadora con arco y flechad",

        "Cazadora",

        "cazadora",

    ],
    "golem_img": [
        "Hombre de piedra",

        "hombre de piedra",

        "Hombre de piedra",

        "Muñeco hombre de piedra",

        "hombre de piedra",
        "Hombre de piedra",

        "Hombre de piedra",

        "Hombre de piedra",
    ],
    "harpy_img": [

        "Figura mitologica",
        "Ser mitologico , gargola ?",
        "Ser mitológico arpía",

        "Animal mitológico",

        "Criatura Mitológica",
        "Ser mitológico",


        "Criatura mitológica",

    ],
    "hellhound_img": [


        "Perro rabioso",

        "Perro",


        "Perro rabioso",


        "Un perro con rabia",
        "Perro",

        "Perro furioso",


        "UN PERRO FURIOSO",


        "Perro diabolico",
    ],
    "hydra_img": [

        "Dragon de 7 cabezas",

        "dragon de siete cabezas",

        "Dragon",

        "Un dragón",
        "dragon",
        "Dragon",

        "dragón de siete cabezas",
        "Dragones",

        "Dragones",
        "Dragón de múltiples cabezas",


        "UN DRAGÓN CON VARIAS CABEZAS",
        "dragon",

        "Dragón de 7 cabezas",
    ],
    "mermaid_img": [
        "Sirena",
        "Imagen de una sirena",
        "Sirena",

        "sirena",
        "Sirena",

        "Sirena",
        "Una sirena",

        "Sirena",


        "Sirena",
        "Una sirena",
        "sirena",
        "Es una sirena",
        "Sirena",
        "sirena",
        "Sirena",
        "Sirena",
        "Sirena",
        "UNA SIRENA",
        "sirena",
        "Sirena, mitico",
        "Sirena",
    ],
    "pegasus_img": [
        "Pegaso",
        "Pegaso mitologia griega",
        "Pegaso",
        "Pegaso",

        "Pegaso",

        "Pegaso",
        "Un pegaso",
        "pegaso",
        "Pegaso",

        "Pegaso",
        "Pegaso",
        "Pegaso",

        "Pegasus",

        "Pegasus",
        "Pegaso",
        "Pegaso",

        "pegasus",
        "Pegaso mítico",

    ],
    "phoenix_img": [

        "El ave fenix",

        "Ave fénix",

        "Phoenix",

        "Ave Fénix",

        "ave fenix",
        "Ave fénix",


        "Ave mítica parecida al fénix de fuego",

        "Ave fenix",


        "Ave Fénix",

    ],
    "thunderbird_img": [
        "Aguila",

        "Águila",
        "Águila",



        "Aguila",


        "Aguila",

        "un águila",

        "Aguila",
        "aguila",
        "Aguila",

    ],
    "triton_img": [
        "Poseidon",

        "Poseidon",

        "Poseidón",
        "Poseidon",
        "poseidon",

        "Poseidón",


        "Poseidon",
        "Poseidón",

        "poseidon",
        "Poseidon",
    ],
    "troll_img": [

        "Ogro ser mitologico",

        "ogro",

        "Un ogro",
        "ogro",

        "Ogro",
        "Ogro",


        "Ogro",
        "un ogro",

        "Ogro",

        "Ogro",

    ],
    "unicorn_img": [
        "Unicornio",

        "Unicornio",
        "Unicornio",
        "unicornio",

        "Unicornio",
        "Unicornio",
        "unicornio",
        "Unicornio",


        "Unicornio",
        "Un unicornio",
        "unicornio",
        "Un unicornio",
        "Unicornio",
        "unicornio",
        "Unicornio",
        "Unicornio",
        "Unicornio",
        "UN UNICORNIO",
        "unicornio",

    ],
    "wyvern_img": [
        "Dragon",

        "Dragón",
        "dragon",
        "Dragón",
        "Dragon",

        "Un dragón",

        "Dragon",
        "Realeza",


        "Un dragón",
        "Dragon",
        "un dragon",
        "Dragon",
        "Dragón",
        "Dragón",
        "UN DRAGÓN",
        "dragon",
        "Dragon",
        "Dragon",
    ],
    "3D_printer_img": [
        "Impresora 3d",

        "Impresora 3D",

        "3D printer",

        "Ompresora 3d",

        "Impresora 3D",

        "Impresora 3D",

        "Impresora 3D",

    ],
    "alienware_desktop_img": [


        "Parlante",
        "Parlante",


        "parlante",


        "Parlante",

        "Un speaker",

        "Parlante",

        "Parlante",

        "Parlante",

    ],
    "alienware_laptop_img": [
        "Laptop",
        "Laptop",

        "Laptop",
        "Laptop",

        "lap top",



        "Laptop",

        "Una laptop",
        "Laptop",

        "Laptop",
        "Laptop",

        "Lap top",
        "Laptop",
        "Laptop",
        "Laptop",
    ],
    "CRT_monitor_img": [



        "Monitor",
        "Monitor antiguo",

        "Monitor antiguo",

        "monitor antiguo",


        "Un monitor",



        "Monitor CRT",

    ],
    "iMac_img": [
        "Pantalla apple",

        "Pantalla mac",

        "Pantalla",


        "Pantalla apple de ordenador",
        "Pantalla",

        "Pantalla Apple",




        "Pantalla",
    ],
    "LCD_monitor_img": [
        "Pantalla de computadora",
        "Pantalla",


        "pantalla",
        "Pantalla de computador",


        "Pantalla computadora",

        "Pantalla",

        "Pantalla",
        "Pantalla",
        "Pantalla",

    ],
    "mac_pro_img": [

        "Parlante",
        "Un parente",


    ],
    "macbook_air_img": [
        "Laptop",
        "Lap top",

        "Laptop",
        "Laptop",
        "laptop",
        "lap top",

        "Laptop",

        "Laptop",
        "Laptop MAC",
        "Una laptop",

        "Laptop",
        "Laptop",

        "Lap top",

        "Laptop",

        "Laptop",
    ],
    "macbook_pro_img": [
        "Laptop",
        "Lap top",

        "Laptop",
        "Laptop",
        "laptop",
        "lap top",
        "Laptop",

        "Laptop",

        "Laptop",

        "Una laptop",

        "Laptop",
        "Laptop",
        "Laptop",

        "Lap top",

        "Laptop",

    ],
    "wired_keyboard_img": [

        "Teclado",
        "Teclado",
        "Teclado",
        "Teclado",
        "teclado",
        "teclado",
        "Teclado",
        "teclado",
        "Teclado",

        "Un teclado",

        "Teclado",

        "Teclado",
        "Teclado",
        "Teclado",
    ],
    "mouse_img": [

        "Mouse",

        "Mouse",

        "mouse",
        "mouse",
        "Mouse",
        "mouse",
        "Mouse",

        "Mouse",

        "Un mouse",
        "Mouse",
        "Mouse",
        "Mouse",
        "Mouse",

        "Mouse",

        "Mouse",
    ],
    "wireless_keyboard_img": [

        "Teclado",

        "Teclado",
        "Teclado",

        "teclado",
        "Teclado",

        "Teclado",

        "Teclado",

        "Un tecldo",

        "Teclado",

        "Teclado",

        "Teclado",

        "Teclado",
        "Teclado",
    ],
    "wireless_mouse_img": [
        "Mouse",
        "Mouse",
        "Ratón",
        "Mouse",

        "mouse",
        "mouse",
        "Mouse",
        "mouse",
        "Mouse",

        "Mouse",

        "Un mouse",
        "Mouse",
        "Mouse",



        "Mouse",
    ],
    "XPS_desktop_img": [
        "CPU",

        "CPU",
        "CPU",

        "CPU (procesador)",


        "Cpu",
        "CPU",
        "Un cpu",

        "CPU",

        "Cpu",

        "Pcu",
    ],
    "XPS_laptop_img": [

        "Lap top",

        "Laptop",
        "Laptop",
        "Laptop",
        "lap top",

        "Laptop",

        "Laptop",
        "Laptop",
        "Una laptop",
        "Laptop",

        "Laptop",
        "Laptop",

        "Lap top",

        "Laptop",
        "Laptop",
    ],
    "chainsaw_img": [

        "Sierra",
        "Sierra",
        "Sierra",

        "motosierra",

        "Sierra",

        "Sierra",

    ],
    "chisel_img": [

        "Herramienta",
        "Herramienta",


        "Alguna herramienta",



        "Herramienta",

        "Herramientas",

        "Herramienta",
    ],
    "drain_shovel_img": [
        "Pala",
        "Pala",
        "Pala",

        "pala",
        "pala",
        "Pala",
        "pala",

        "Pala",

        "Pala",

        "Pala",
        "Pala",

        "Una pala",
        "Pala",
    ],
    "electric_pruner_img": [



        "Tijera",




        "Tijera de podar",
        "Tijeras",

        "Tijetas",

        "Tijera",

    ],
    "flat_brush_img": [

        "Pincel",


        "Pincel",

        "pincel",
        "pincel",
        "Pincel",
        "pincel",



        "Pincel",

        "Pincel",
        "Pincel",

        "Pincel",
    ],
    "gardening_shear_img": [
        "Tijera de jardinería",


        "Tijera de jardin",

        "tijera de jardinería",




        "Tijera para jardinería",


        "Tijeras para jardinería",
        "Tijera jardinería",
        "Tijeras jardinería",


    ],
    "glass_palette_img": [

        "Paleta con colores",

        "Paleta de colores",

        "paleta de colores",

        "Paleta de colores para pintar cuadros",


        "Una paleta de colores",

        "Paleta de colores de acuarela",
        "Paleta de colores",

    ],
    "lopper_img": [

        "Alicates",
        "Alicate",

        "alicate grande",

        "Alicate",

    ],
    "mallet_img": [

        "Martillo para cocinar",
        "Martillo",
        "Martillo",



        "Martillo para comida",




        "Martillo para Carne",


        "Martillo",

        "Martillo",

    ],
    "potting_trowel_img": [
        "Pala para jardinería",

        "Pala jardinería",

        "Una Pala de jardineria",




        "Palita jardinería",



        "Pala jardinería",


        "Pala de jardinería",


        "Pala de Jardineria",
    ],
    "rake_img": [

        "Rastrillo",
        "Rastrillo",
        "Rastrillo",

        "rastrillo",
        "rastrillo",

        "rastrillo",
        "Rastrillo",
        "Rastrillo playa",
        "Rastrillo",
        "Rastrillo",

        "Rastrillo",
        "Rastrillo",
        "Rastrillo",



        "Rastrillo",
    ],
    "rockery_trowel_img": [

        "Herramienta",
        "Una herramienta para jardin",

        "herramienta de jardinería",


        "Herramienta",

        "Herramienta",

    ],
    "round_brush_img": [
        "Pincel",
        "Pincel",

        "Pincel",
        "Pincel",
        "princel",
        "pincel",
        "Pincel",
        "pincel",

        "Pincel",
        "Pincel",

        "Un pincel",

        "Pincel",
        "Pincel",
        "Pincel",
        "Pincel",
        "Pincel",
        "Pincel",
        "Pincel",
        "Pincel",
    ],
    "rounded_point_shovel_img": [
        "Pala",
        "Pala",
        "Pala",
        "Pala",

        "pala",
        "pala",
        "Pala",
        "pala",

        "Pala",
        "Pala",
        "Una pala",
        "Pala",

        "Pala",
        "Pala",
        "Pala",
        "Pala",


        "Pala",
    ],
    "wire_img": [
        "Cable",
        "Cable",

        "Cable",

        "Cable",

        "Cable enrollado",
        "Cables",




        "Cable",


        "Cable",
    ],
    "wood_palette_img": [
        "Paleta para mezclar pintura",
        "Paleta para poner pinturas",


        "Paleta de pinturas",


        "Paleta pintura",
        "Paleta pintura",




        "Paleta para pintura",




    ],
    "Brocolli_img": [
        "Brócoli",

        "Brocoli",
        "Brocoli",
        "Brócoli",

        "brócoli",
        "Brocoli",
        "Brócoli",
        "Brocolis",


        "Brócoli",
        "Brócoli fresco crudo",
        "Brócoli",
        "Brócoli",
        "Brocoli",

        "Brócoli",
        "Brocoli",

        "brocoli",
        "Brocoli",
    ],
    "Carrot_img": [
        "Zanahorias",
        "Fresca zanahoria",
        "Zanahoria",
        "Zanahoria",
        "Zanahoria",

        "zanahorias",
        "Zanahorias",
        "Zanahorias",
        "Zanahorias",

        "Zanahoria",
        "Zanahoria",

        "Zanahorias",
        "Zanahorias",
        "Zanahoria",
        "Zanahoria",
        "Zanahoria",
        "Zanahoria verdura",

        "Zanahoria",
        "zanahorias",
        "Zanahorias",
    ],
    "Cauliflower_img": [
        "Coliflor",

        "Coliflor",
        "Coliflor",
        "Coliflor",

        "coliflor",
        "Coliflor",
        "Coliflor",
        "Coliflor",
        "Coliflor",
        "Coliflor",
        "Coliflor",

        "Coliflor",
        "Coliflor",
        "Coliflor",
        "Coliflor",
        "Coliflor",
        "Vegetal",
        "Coliflor",
        "Coliflor",
        "coliflor",
    ],
    "Chicken_Breast_img": [

        "Pechuga de pollo",

        "Pechuga de pollo",
        "Pechuga de pollo",

        "Filete de pechuga",

        "Pechuga",
        "Pechuga de pollo",
        "Pechuga pollo",

        "Pechuga pollo",


    ],
    "Apple_img": [

        "Manzana verde",
        "Manzana verde",

        "Manzana verde",

        "manzana verde",
        "Manzana verde",

        "Manzana verde",

        "Manzana verde entera y partida",

        "Manzana verde",

        "Manzana verde",
        "manzana verde",
    ],
    "Ham_slice_img": [
        "Jamón",

        "Jamon",

        "Jamón",


        "Jamón",
        "Jamón",

        "Jamon",
        "Jamón",
        "Jamón",
        "Jamón",
        "Jamón",
        "Jamon",
        "Jamón",
        "Jamón",

    ],
    "McIntosh_Apple_img": [
        "Manzana",

        "Manzana",
        "Manzana",

        "Manzana",
        "Manzana",

        "Manzana",
        "Manzana",
        "Manzana",

        "Manzana",
        "Manzana",
        "Manzana",
        "Manzana",
        "Manzana",
        "Manzana",

    ],
    "moon_drop_grape_img": [
        "Uvas",

        "Uvas",

        "uvas",
        "uvas",

        "Uvas",


        "Uvas",

        "Uvas",
        "Uva",

        "Fruta",

    ],
    "pork_sausage_img": [
        "Chorizos",

        "Chorizos",
        "Chorizo",
        "Chorizos",

        "chorizo",
        "Chorizos",

        "Chorizo",
        "Chorizos",
        "Chorizo",
        "Chorizo",

        "Chorizos",

        "Chorizo",
        "Chorizos",
        "Chorizos",
        "Chorizo",

        "Chorizo",
        "Chorizos",
    ],
    "red_globe_grape_img": [
        "Uvas",

        "Uvas",
        "Uvas",

        "uvas",
        "Uvas",
        "Uvas",

        "Uvas",
        "Uvas",

        "Uvas",
        "Uvas",

        "Uvas",
        "Uvas",
        "Uvas",

    ],
    "ribeye_steak_img": [

        "Carne de res",
        "Carne",


        "carne cruda",

        "Carne",
        "Carne",
        "Pedazo de carne",
        "Carne res",
        "Carne",

        "Carne",

        "Carne",



    ],
    "salmon_fillet_img": [

        "Salmón",
        "Salmon",
        "Salmón",

        "salmon",
        "Salmon",
        "Salmón",
        "Salmón",

        "Salmon",
        "Salmón",

        "Salmon",
        "Salmón",

        "Salmón",

        "Salmón crudo",
        "Salmón",
    ],
    "sweet_potato_img": [
        "Camote",

        "Camote",
        "Camote",
        "Camotes",
        "Camotes",
        "camotes",

        "Camote amarillo",
        "Camote",
        "Camote",
        "Camote",

        "Camotes",
        "Camote",
        "Camote",

        "Camote",

    ],
    "t_bone_steak_img": [
        "T bone steak",

        "T bone",

        "carne t bone",


        "T bone",

        "T bone steak",

        "T bone",

        "Carne T-bone",
        "T-bone",

    ],
    "tuna_steak_img": [
        "Atún",



        "Atún",
        "Atún",
        "Atun",
        "Atun",
        "Atun",

        "Atun",
        "Atún rojo",
        "Atun",

        "Atún",
    ],
    "turkey_breast_img": [

        "Pavo",

        "Pavo",
        "Pavo",
        "Pavo",

        "Pavo",
        "Pavo",

    ],
    "bass_trombone_img": [
        "Trompeta",

        "Trompeta",
        "Trompeta",
        "Trompeta",

        "trompeta",

        "Trompeta",

        "Trompeta",
        "Trompeta",
        "Trompeta",

        "Trompeta",
        "Trompeta",
        "Trompeta",

        "Trompeta",

        "Trompeta",
        "trompeta",

    ],
    "bell_img": [
        "Campana",

        "Campana",
        "Campana",
        "Campana",

        "campana",
        "Campana",
        "Campana",
        "Campana",
        "Campana",
        "Campana",
        "Campana",

        "Campana",
        "Campana",
        "Campana",
        "Campana",
        "Campana",
        "Campana",
        "Campana",
        "Campana",
    ],
    "bongo_drums_img": [

        "Tambores",
        "Tambores",

        "Tambores",


        "Tambores",
        "Tambores",

        "Tambores",



        "Tambores",

        "Tambores. Tumba?",

    ],
    "castanets_img": [
        "Castañuelas",

        "Castañuela",

        "Castañuelas",


        "Castañuelas",
        "Castañuelas",
        "Castañuelas",
        "Castañuelas",
        "Cadtañuela",
        "Castañuelas",

        "Castañuelas",
        "Castañuelas",
        "Castañuelas",
        "Castañuelas",
        "Castañuelas",

        "Castañuelas",
        "cascañuelas",
        "Castañuelas",
    ],
    "clarinet_img": [

        "Clarinete",
        "Clarinete",

        "Clarinete",
        "clarinete",
        "Clarinete",



        "Clarinete",
        "Clarinete",
        "Clarinet",
        "Clarinete",
        "Clarinete",

        "Clarinete",
    ],
    "cornet_img": [
        "Trompeta",

        "Trompeta",

        "Trompeta",

        "trompeta",
        "Trompeta",
        "Trompeta",
        "Trompeta",
        "Trompeta",
        "Trompeta",
        "Trompeta",
        "Trompeta",
        "Trompeta",

        "Trompeta",

    ],
    "fife_img": [
        "Flauta",

        "flauta",
        "Flauta",
        "Flautin",
        "Flauta",
        "Flauta",
        "Flauta",
        "Flauta",

        "Flauta",

        "Flauta",
        "Flauta",

    ],
    "piccolo_img": [
        "Flauta",

        "Flauta traversa",
        "Flauta",


        "Flauta",
        "Flauta",
        "Oboe",

        "Flauta traversa",
        "Flauta",

        "Flauta",

    ],
    "roto_drums_img": [

        "Tambor",

        "Tambores",


        "Tambores",


        "Tambores",
        "Tambores",
        "tambores",

    ],
    "saxophone_img": [
        "Saxofón",

        "Saxo",
        "Saxofone",

        "saxofón",
        "Saxofón",
        "Saxofón",
        "Saxo",

        "Saxofón",
        "Saxofón",
        "Saxo",
        "Saxofón",
        "Saxofon",

        "Saxofón",

        "Saxofon",
    ],
    "slit_drum_img": [
        "Xilófono",

        "Xilofono",


        "xilófono",

        "Xilifono",
        "Xilofono",
        "Xilofon",
        "Xilófono",
        "Xylofono",
        "Xilofono",
        "Xilófono",

        "xilófono",
    ],
    "tambourine_img": [

        "Pandereta",
        "Pandereta",
        "Pandereta",

        "pandereta",
        "Pandereta",
        "Pandereta",
        "Pandereta",

        "Pandereta",
        "Pandereta",
        "Pandereta",
        "Pandereta",
        "Pandereta",
        "Pandereta",
        "Pandereta",
        "Pandereta",
        "Una pandereta",
        "Pandereta",

        "pandereta",

    ],
    "tenor_trombone_img": [
        "Trompeta",
        "Trompeta",
        "Trompeta",
        "Trompeta",
        "Trompeta",

        "Teompeta",
        "Trompeta",
        "Trompeta",

        "Trompeta",

        "Trompeta",

        "Triompeta",
        "Trompeta",
        "trompeta",

    ],
    "timpani_img": [



        "Tambor",

        "Tambor",

        "Tambor",

        "Especie de tambor",
        "tambor",
    ],
    "triangle_img": [

        "Triangulo",
        "Triangulo",
        "Triángulo",

        "triángulo",
        "Triangulo",

        "Triangulo",
        "Triangulo musical",
        "Triangulo",
        "Triángulo",

        "Triángulo",
        "Triangulo",
        "Triangulo",
        "Triángulo musical",
        "Triangulo",
        "Triangulo",

        "Triángulo",
    ],
    "trumpet_img": [
        "Trompeta",

        "Trompeta",
        "Trompeta",
        "Trompeta",
        "Trompeta",
        "trompeta",
        "Trompeta",
        "Trompeta",
        "Trompeta",
        "Teompeta",
        "Trompeta",
        "Trompeta",
        "Trompeta",
        "Trompeta",

        "Trompeta",
        "Trompeta",
        "Trompeta",

        "Trompeta",
        "trompeta",
    ],
}







# =========================
# Normalization
# =========================
# Clean the dataframe labels
df['Image'] = df['Image'].str.strip().str.lower()

# Clean the correct_answers dictionary
correct_answers_normalized = {
    k.lower().strip(): [ans.lower().strip() for ans in v]
    for k, v in correct_answers.items()
}

# =========================
# Agreement Rate Calculation with Debugging
# =========================
image_agreement = {}
missing_images = []
unmatched_responses = {}

for image in df['Image'].unique():
    if image not in correct_answers_normalized:
        missing_images.append(image)
        continue

    image_df = df[df['Image'] == image]
    responses = image_df['Response'].dropna().str.strip().str.lower()

    acceptable_answers = correct_answers_normalized[image]

    agree_count = 0
    disagree_count = 0
    unmatched_resp = []

    for resp in responses:
        if resp in acceptable_answers:
            agree_count += 1
        else:
            disagree_count += 1
            unmatched_resp.append(resp)

    total = agree_count + disagree_count
    agreement_rate = (agree_count / total) * 100 if total > 0 else 0

    image_agreement[image] = {
        'Agreement %': agreement_rate,
        'Agree Count': agree_count,
        'Disagree Count': disagree_count,
        'Total Responses': total
    }

    if unmatched_resp:
        unmatched_responses[image] = Counter(unmatched_resp)

# =========================
# Chi-Square Test
# =========================
# Prepare contingency table
contingency_data = []
labels = []

for image, data in image_agreement.items():
    if data['Total Responses'] >= 5:  # Ensure enough sample size
        contingency_data.append([data['Agree Count'], data['Disagree Count']])
        labels.append(image)

# Run chi-square if we have enough images
if len(contingency_data) >= 2:
    chi2, p, dof, expected = chi2_contingency(contingency_data)

    print("\nChi-Square Test Results")
    print(f"Chi2 Statistic: {chi2}")
    print(f"p-value: {p}")
    print(f"Degrees of Freedom: {dof}")
    print(f"Expected Frequencies:\n{expected}")

    # Interpret
    alpha = 0.05
    if p < alpha:
        print("\nConclusion: There is a statistically significant difference in agreement rates across images (reject null hypothesis).")
    else:
        print("\nConclusion: No significant difference in agreement rates across images (fail to reject null hypothesis).")
else:
    print("\nNot enough images with sufficient responses to perform chi-square test.")

# =========================
# Outputs
# =========================
print("\nAgreement Rates per Image:")
for image, stats in image_agreement.items():
    print(f"{image}: {stats['Agreement %']:.2f}% (n={stats['Total Responses']})")

if missing_images:
    print("\nMissing Images from Dictionary:")
    print(missing_images)

if unmatched_responses:
    print("\nUnmatched Responses by Image:")
    for image, counts in unmatched_responses.items():
        print(f"{image}: {dict(counts)}")
```
