```python
import pandas as pd
from collections import Counter
from scipy.stats import chi2_contingency
import numpy as np

# Load your Excel file (replace the path with your actual file)
file_path = '/content/sample_data/NONEXPERTS_NL_CLEAN (1).xlsx'
df = pd.read_excel(file_path)

# Example correct answers dictionary
correct_answers = {
    "Brocolli_img": [
        "Brocoli",
        "Broccoli",
        "Broccoli",
        "Broccoli",
        "broccoli",
        "Broccoli",
        "Broccoli",
        "Broccoli",
        "broccoli",

        "broccoli",
        "Broccoli",
        "Broccoli",

    ],
    "Carrot_img": [
        "Wortels",

        "wortel",
        "Wortels",

        "Wortels",

    ],
    "Cauliflower_img": [
        "Bloemkool",
        "Bloemkool",
        "Bloemkool",
        "Bloemkool",
        "bloemkool",
        "Bloemkool",
        "Bloemkool",
        "Bloemkool",
        "bloemkool",
        "bloemkool",
        "Bloemkool",
        "bloemkool.",
        "bloemkool",
        "Bloemkool",
    ],
    "Chicken_Breast_img": [

        "Kipfilet",

        "Kipfilet",

        "Kipfilet",

        "kipfilet",
        "Kipfilets",
        "Kip filet",
    ],
    "Apple_img": [
        "Appel",


        "Appel",

        "Appel - Granny Smith",

        "Appel",

        "appel",

        "appel",
        "Appels",

    ],
    "Ham_slice_img": [
        "Ham",

        "Ham",

        "ham",

        "ham",
        "Plak ham",
    ],
    "McIntosh_Apple_img": [
        "Appel",
        "Appel",
        "Appel",
        "Appel",
        "appel",
        "Appel",
        "Appel",
        "Appel",

        "appel",
        "Appel",
    ],
    "moon_drop_grape_img": [

        "Druiven",
        "Druiven",

        "druiven",
        "Druiven",

        "druiven",

    ],
    "pork_sausage_img": [

        "Worst",
        "Worst",
        "Worst",

        "Worst",

        "worstjes",

        "worstjes",
        "Worstjes",
    ],
    "red_globe_grape_img": [

        "Tros druiven",

        "Tros rode druiven",
        "Tros druiven",
        "tros druiven",

        "een tros druiven",

        "Tros druiven",
    ],
    "ribeye_steak_img": [
        "Vlees",
        "Vlees",
        "Rundvlees",




        "rundvlees (entrecote?)",

    ],
    "salmon_fillet_img": [
        "Zalm",


        "zalm",
        "Zalm",




    ],
    "sweet_potato_img": [

        "Zoete aardappel",
        "zoete aardappel",

        "Zoete aardappel",
        "Zoete aardappel",
        "zoete aardappelen",
        "zoete aardappelen",

        "zoete aardappel",

    ],
    "t_bone_steak_img": [

        "T-bone steak",
        "T-bone",
        "t-bone steak",

        "T-bone steak",


        "rundvlees (T-bone?)",
        "T-bone steak",
    ],
    "tuna_steak_img": [

        "Tonijn",
        "tonijn",
        "Tonijn",
        "Stuk tonijn",

    ],
    "turkey_breast_img": [

        "gebakken kipfilet",


        "gebakken kipfilet",
        "Gebakken kipfilet",

    ],
    "bass_trombone_img": [

        "Trompet",

        "Trompet",

        "schuiftrompet",
        "Trompet",

        "trompet",

    ],
    "bell_img": [
        "Bel",
        "Bel",
        "Bel",
        "Bel",

        "Bel",
        "Bel",
        "Bel",

        "bel",
        "Een bel",

        "bel",

    ],
    "bongo_drums_img": [



        "trommel",
        "Trommels",
        "twee trommels",
        "trommels",
        "Trommels",

    ],
    "castanets_img": [

        "Clarinettes",
        "Castagnette",
        "castagnetten",
        "Castagnettes",
        "Castagnetten",
        "Castagnetten",
        "castagnetten",

        "castannetes",
        "Castanets",

        "Castagnetten",
    ],
    "clarinet_img": [
        "Klarinet",

        "klarinet",

        "Clarinet",
        "Klarinet",

        "klarinet",
        "klarinet",

        "clarinet",
        "Klarinet",
        "Klarinet",

    ],
    "cornet_img": [
        "Trombone",
        "Trompet",

        "Trompet",
        "Soort van trompet",


        "trompet-achtige",
        "Trompet",
    ],
    "fife_img": [

        "Fluit",
        "Fluit",

        "Fluit",
        "Fluit"
        "blokfuit"

        "Blokfluit",

        "fluit",
        "Blokfuit",
    ],
    "piccolo_img": [

        "Fluit",
        "Fluit",
        "Fluit",

        "Fluit",

        "fluit",

    ],
    "roto_drums_img": [
        "Drumstel",
        "Drum stel",
        "Drum",
        "Drums",

        "Drumstel",
        "Drumstel",
        "Drumset",
        "drums",
        "drumstel",

        "Drum soort",
    ],
    "saxophone_img": [
        "Saxefoon",
        "Saxofoon",
        "Saxofoon",
        "Saxafoon",
        "saxofoon",
        "Saxofoon",

        "Saxofoon",

        "saxofoon",
        "saxofoon",

        "saxofoon",
        "Saxophone",
    ],
    "slit_drum_img": [

        "Xylofoon",
        "xylofoon",

        "Xylofoon met houten tongboxen",


        "Een xylofoon",

        "xylofoon-achtige",
        "Xylophone",
    ],
    "tambourine_img": [


        "tamboerijn",

        "Tamboerijn",
        "Tamboerijn",
        "tamboerijn",
        "tamborijn",
        "Tamboerijn",
        "tamboerijn",
        "Tamboerijn",
    ],
    "tenor_trombone_img": [



        "Trombone",
        "trombone",

        "Trombone of schuiftrompet",
        "Trombone",
        "trombone",

        "trombone",
        "ik zie een koper blaasinstrument. Een trombone o.i.d.",

        "Trombone",
    ],
    "timpani_img": [

        "Pauk",

        "Pauk",
        "Pauk",

        "pauke",


        "pauk",
        "Pauk",

    ],
    "triangle_img": [
        "Triangel",

        "Triangel",
        "Triangle",
        "triangel",
        "Triangel",
        "Triangel",
        "Triangel",
        "triangel",
        "triangel",
        "triangel",
        "ik zie een triangel.",
        "triangel",
        "Triangel",
        "Triangel",
    ],
    "trumpet_img": [
        "Trompet",
        "Trompet",

        "Trompet",
        "trompet",
        "Trompet",
        "Trompet",
        "Trompet",
        "trompet",
        "trompet",
        "Trompet",
        "trompet",
        "trompet",
        "Trompet",
        "Trumpet",
    ],
    "3D_printer_img": [
        "3D-printer",

        "3D printer",
        "3D printer",

        "3D-printer",

        "3D-Printer",
        "3d printer?",
        "3d-printer??",
        "3D printer",
        "3D printer",
    ],
    "alienware_desktop_img": [

        "speaker?",

        "Speaker",
        "luidspreker?",

        "speaker",

    ],
    "alienware_laptop_img": [

        "Laptop",
        "Laptop",
        "Laptop",

        "Laptop",
        "laptop",
        "Laptop",

    ],
    "CRT_monitor_img": [

        "oude computer",

        "Computer",

        "Computer",

    ],
    "iMac_img": [
        "iMac",

        "Imac",

        "Mac",

        "Imac",
        "Mac beeldscherm",
        "mac-scherm",


    ],
    "LCD_monitor_img": [

        "Beeldscherm",

        "Beeldscherm",
        "beeldscherm",
        "Monitor of Beeldscherm",
        "computer beeldscherm",
        "computerscherm",

    ],
    "mac_pro_img": [

        "Rasper",
        "Rasp",

    ],
    "macbook_air_img": [
        "laptop",
        "laptop",

        "Laptop",
        "Laptop",
        "Laptop",
        "Laptop",
        "laptop",
        "Laptop",
        "laptop",
        "laptop",
        "Laptop",

    ],
    "macbook_pro_img": [
        "laptop",
        "laptop",
        "Laptop",
        "Laptop",
        "Laptop",
        "Laptop",
        "Laptop",
        "laptop",
        "Laptop",
        "laptop",
        "laptop",
        "Laptop",

    ],
    "wired_keyboard_img": [
        "toetsenbord",
        "keyboard",
        "Keyboard",
        "Keyboard",
        "Toetsenbord",
        "Toetsenbord",
        "Toetsenbord",

        "Toetsenbord",

        "toetsenbord",

    ],
    "mouse_img": [

        "muis",
        "Muis",

        "Muis",


        "Muis",

        "Muis",

        "muis",

        "Muis",
        "Muis",
    ],
    "wireless_keyboard_img": [

        "keyboard",
        "Keyboard",
        "Keyboard",
        "Toetsenbord",

        "Toetsenbord",
        "toetsenbord",
        "Toetsenbord",
        "toetsenbord",

        "Toetsenbord",


    ],
    "wireless_mouse_img": [

        "muis",
        "Muis",
        "Muis",
        "Muis",
        "Muis",
        "Muis",

        "Muis",


    ],
    "XPS_desktop_img": [

        "Computerkast",
        "Computerkast",
        "Computerkast",


        "computerkast / desktop",

    ],
    "XPS_laptop_img": [
        "laptop",
        "laptop",
        "Laptop",
        "Laptop",
        "Laptop",

        "Laptop",
        "laptop",
        "Laptop",
        "laptop",

        "Laptop",

    ],
    "chainsaw_img": [
        "kettingzaag",
        "chainsaw",

        "Kettingzaag",

        "kettingzaag",
        "Kettingzaag",
        "kettingzaag",
        "kettingzaag",
        "Kettingzaag",
        "Kettingzaag",
    ],
    "chisel_img": [


        "Houtenbeitel",
        "Beitel",

        "Beitel",

    ],
    "drain_shovel_img": [
        "spade",
        "schop",
        "Schop",


        "tuinschop",

        "Schop",
        "Schep",
    ],
    "electric_pruner_img": [
        "snoeischaar",
        "snoeischaat",

        "Snoeischaar",

        "Snoeischaar",


        "snoeischaar",
        "snoeischaartje",

    ],
    "flat_brush_img": [
        "kwast",
        "kwast",

        "Kwast",

        "Kwast",
        "kwastje",
        "Kwast",
        "penseel of kwast",
        "kwastje",
        "Kwast",
        "Kwast",

    ],
    "gardening_shear_img": [
        "snoeischaar",
        "snoeischaar",

        "Snoeischaar",

        "Snoeischaar",

        "snoeischaar",

        "snoeischaar",
        "snoeischaartje",

    ],
    "glass_palette_img": [

        "Verf Palette",

        "Verfpalet",


        "Verf palet met verf",
        "Verf palet",
    ],
    "lopper_img": [

        "snoeischaar",

        "Snoeischaar",

        "Snoeischaar",


        "snoeischaar met lange arm",

    ],
    "mallet_img": [

        "meat masher",
        "Vlees hamer",
        "Vlees hamer",

        "Vleeshamer",
        "vleeshouwer",

        "Vleeshamer",

        "Vlleshamer",
    ],
    "potting_trowel_img": [
        "gardening tool",

        "Schep",
        "Schep",
        "Tuinschepje",
        "tuinschepje",
        "Schepje",

        "tuinschepje",


    ],
    "rake_img": [
        "hark",
        "hark",
        "Hark",

        "Hark",
        "Hark",
        "Hark",

        "Hark",


        "Hark",
        "hark",
    ],
    "rockery_trowel_img": [



        "tuin gereedschap??",
        "Tuinier gereedschap voor zaadjes",

    ],
    "round_brush_img": [
        "kwast",
        "kwast",
        "Verf kwast",

        "Kwast",

        "Kwast",

        "Kwast",
        "Kwast",
    ],
    "rounded_point_shovel_img": [
        "schop",
        "schop",
        "Schop",
        "Delf schop",

        "Spade",
        "Spade",


        "schop",
        "Schop",

        "Schop",
    ],
    "wire_img": [


        "Ijzerdraad",
        "IJzerdraad",

        "Ijzerdraad",

        "rol ijzerdraad",
        "Ijzer draad",
    ],
    "wood_palette_img": [



        "Verfpalet",
        "verfpalet",

        "Verf palet",
        "Verf palet",

    ],
    "apple_tree_img": [

        "Appel boom",
        "appelboom",
        "Appelboom",
        "Appelboom",
        "Appelboom",

        "Appelboom",
        "Appelboom",

        "Appelboom",
        "appel boom",
        "Appelboom",
        "Appel boom",
    ],
    "bulldog_img": [

        "Buldog",
        "Bulldog",
        "bulldog",
        "Buldog",
        "Bulldog",
        "Engelse buldog",
        "Buldog, hond",
        "Hond buldog",
        "bulldog",
        "Bulldog",

    ],
    "cedar_img": [

        "Boom",

        "Boom",


        "Boom",
        "boom",
        "Een boom",
    ],
    "dandelion_img": [

        "Paardenbloem",


        "Paardebloem",


        "paardebloem",
        "Paardenbloem",

    ],
    "Eagle_img": [

        "Arend/ vogel",
        "Eagle",
        "eagle",
        "Adelaar",

        "Eagle",
        "Arend",
        "Eagle",
        "adelaar",
        "Arend",
        "adelaar",
        "Adelaar",
        "Eagle",
    ],
    "Edelweiss_img": [
        "Flower?",

        "Ijsbloem",
        "bloem",
        "Bloem",



        "bloem",

        "Bloem",
    ],
    "german_shepherd_img": [
        "German shepherd",
        "Duitse herdershond",
        "hond (german shepherd)",
        "Herdershond",
        "Herdershond",
        "Duitse herder",
        "Duitse herdershond",

        "Herdershond",
        "herdershond",

    ],
    "orange_tree_img": [
        "Sinaasappel boom",
        "sinaasappel boom",
        "Sinaasappelboom",
        "Sinaasappelboom",
        "Sinaasappelboom",
        "Sinaasappel boom",
        "Sinaasappelboom",
        "Sinaasappel boom",
        "sinasappelboom",
        "Sinaasappelboom",
        "sinaasappel boom",
        "Sinaasappelbopm",
        "Sinaasappel boom",
    ],
    "Ostrich_img": [
        "Struisvogel",
        "Struisvogel",
        "Struisvogel",
        "Struisvogel",
        "Struisvogel",
        "Struisvogel",
        "Struisvogel",
        "Struisvogel",
        "Struisvogel",
        "struisvogel",
        "Struisvogel",
        "Struisvogel",
        "struisvogel",
        "Struisvogel",
        "Struisvogel",
    ],
    "owl_img": [
        "Owl",
        "Uil",
        "Uil",
        "uil",
        "Uil",
        "Uil",
        "Uil",
        "Uil",
        "Uil",
        "Uil",
        "uil",
        "uil",
        "Uil",
        "uil",
        "Uil",
        "Een uil",
    ],
    "penguin_img": [
        "Penguins",
        "Pinguin",
        "Penguin",
        "pinguïn",
        "Pinguin",
        "Pinguïn",
        "Penguin",
        "Pinguïn",
        "Pinguïn",
        "pinguin",
        "pinguin",
        "Pinguin",
        "pinguin",
        "Pinguin",
        "Penguin",
    ],
    "persian_img": [
        "Cat",

        "Kat",


        "Kat",


        "kat",

        "kat",
        "Kat",
        "Kat",
    ],
    "Pine_img": [

        "Boom",

        "boom",
        "Boom",

        "Boom",

        "boom",
        "Boom",
        "boom",
        "Boom",
    ],
    "rose_img": [
        "Rose",

        "roos",
        "Roos",

        "Roos",
        "Roos",
        "Roos",
        "Rios",
        "roos",

        "roos",
        "roos",
        "Roos",
        "Rose",
    ],
    "siamese_img": [
        "Cat",
        "Kat",
        "Kat",
        "kat",

        "Kat",
        "Kat",

        "kat",

        "Kat",
    ],
    "tulip_img": [
        "Tulip",

        "Tulp",

        "Tulp",
        "Tulp",

        "Tulp",
        "Tulp",
        "Tulp",
        "tulp",
        "Bloem - tulp",


        "Tulp",
        "Tulip",
    ],
    "cerberus_img": [


        "Driekoppige hond",
        "Driekoppige hond",


        "Driekoppige draak",

        "driekoppige hond",
        "Driekoppige hond",

    ],
    "cherubim_img": [


        "Engel",
        "engeltje",
        "Engeltje",
        "Engeltje",


        "Engel",



        "engel",
        "Engel",
        "Engel",
    ],
    "drow_img": [




        "Elf",

        "elf",
        "Elf",

    ],
    "forest_elf_img": [


        "Boogschutter",
        "boogschutter in het bos",
        "Boogschutter",
        "Bos boogschutter",

        "boogschutter",

    ],
    "golem_img": [

        "Golem",

        "Golem",

        "Golem",

    ],
    "harpy_img": [

        "Vouwvogel figuur",

        "vogelvrouw",

    ],
    "hellhound_img": [

        "Hellehond",
        "Hellehond",
        "Hellehond",

        "Hell hond",

    ],
    "hydra_img": [

        "Hydra",


        "Hydra",

        "Hydra",
        "Hydra",
    ],
    "mermaid_img": [
        "Zeemeermin",
        "Zeemeermin",
        "Zeemeermin",
        "mermaid",
        "Zeemeermin",
        "Zeemeermin",
        "Zeemeermin",
        "Meermin",
        "Zeemeermin",
        "Zeemeermin",
        "zeemeermin",
        "zeemeermin",
        "Mermaid",
        "zeemeermin",
        "Zeemeermin",
        "Zeemeermin",
    ],
    "pegasus_img": [

        "Pegasus",
        "pegasus",
        "Pegasus",

        "Pegasus",

        "Pegasus",
        "Pegasus",
        "pegasus",

        "Pegasus",

        "Pegasus",
        "Pegasus",
    ],
    "phoenix_img": [

        "Phoenix",

        "Phoenix",
        "Phoenix",


        "Fenix",

        "Feniks",
        "Fenix",
        "Feniks",
    ],
    "thunderbird_img": [


        "eagle",
        "Adelaar",
        "Getekende adelaar",

        "Adelaar symbool",



        "Adelaar",
        "adelaar",


    ],
    "triton_img": [

        "Poseidon",

        "Poseidon",


        "Tekening - Poseidon met drietand",
        "Poseidon",
        "Poseidon",
        "Poseidon",

    ],
    "troll_img": [

        "Trol",
        "Trol",
        "Trol",

        "Troll",

        "Trol",

        "Troll",
        "Troll",
    ],
    "unicorn_img": [
        "Unicorn",

        "Unicorn",
        "unicorn",
        "Eenhoorn",

        "Unicorn",
        "Eenhoorn",
        "Eenhoorn",
        "Unicorn",
        "eenhoorn",

        "Eenhoorn",
        "eenhoorn",
        "Eenhoorn",
        "Unicorn",
    ],
    "wyvern_img": [
        "Draak",
        "Draak",
        "Draak",
        "draak",
        "Draak",

        "Draak",
        "Draak",


        "Draak",
        "draak",
        "Draak",

    ],
}









# =========================
# Normalization
# =========================
# Clean the dataframe labels
df['Image'] = df['Image'].str.strip().str.lower()

# Clean the correct_answers dictionary
correct_answers_normalized = {
    k.lower().strip(): [ans.lower().strip() for ans in v]
    for k, v in correct_answers.items()
}

# =========================
# Agreement Rate Calculation with Debugging
# =========================
image_agreement = {}
missing_images = []
unmatched_responses = {}

for image in df['Image'].unique():
    if image not in correct_answers_normalized:
        missing_images.append(image)
        continue

    image_df = df[df['Image'] == image]
    responses = image_df['Response'].dropna().str.strip().str.lower()

    acceptable_answers = correct_answers_normalized[image]

    agree_count = 0
    disagree_count = 0
    unmatched_resp = []

    for resp in responses:
        if resp in acceptable_answers:
            agree_count += 1
        else:
            disagree_count += 1
            unmatched_resp.append(resp)

    total = agree_count + disagree_count
    agreement_rate = (agree_count / total) * 100 if total > 0 else 0

    image_agreement[image] = {
        'Agreement %': agreement_rate,
        'Agree Count': agree_count,
        'Disagree Count': disagree_count,
        'Total Responses': total
    }

    if unmatched_resp:
        unmatched_responses[image] = Counter(unmatched_resp)

# =========================
# Chi-Square Test
# =========================
# Prepare contingency table
contingency_data = []
labels = []

for image, data in image_agreement.items():
    if data['Total Responses'] >= 5:  # Ensure enough sample size
        contingency_data.append([data['Agree Count'], data['Disagree Count']])
        labels.append(image)

# Run chi-square if we have enough images
if len(contingency_data) >= 2:
    chi2, p, dof, expected = chi2_contingency(contingency_data)

    print("\nChi-Square Test Results")
    print(f"Chi2 Statistic: {chi2}")
    print(f"p-value: {p}")
    print(f"Degrees of Freedom: {dof}")
    print(f"Expected Frequencies:\n{expected}")

    # Interpret
    alpha = 0.05
    if p < alpha:
        print("\nConclusion: There is a statistically significant difference in agreement rates across images (reject null hypothesis).")
    else:
        print("\nConclusion: No significant difference in agreement rates across images (fail to reject null hypothesis).")
else:
    print("\nNot enough images with sufficient responses to perform chi-square test.")

# =========================
# Outputs
# =========================
print("\nAgreement Rates per Image:")
for image, stats in image_agreement.items():
    print(f"{image}: {stats['Agreement %']:.2f}% (n={stats['Total Responses']})")

if missing_images:
    print("\nMissing Images from Dictionary:")
    print(missing_images)

if unmatched_responses:
    print("\nUnmatched Responses by Image:")
    for image, counts in unmatched_responses.items():
        print(f"{image}: {dict(counts)}")
```
