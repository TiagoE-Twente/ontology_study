```python
import pandas as pd
from collections import Counter
from scipy.stats import chi2_contingency
import numpy as np

# Load your Excel file (replace the path with your actual file)
file_path = '/content/sample_data/NONEXPERTS_IT_CLEAN (2).xlsx'
df = pd.read_excel(file_path)

# Example correct answers dictionary
correct_answers  = {
    "apple_tree_img": [

    ],
    "bulldog_img": [

        "cane",
        "Il cane",

    ],
    "cedar_img": [

    ],
    "dandelion_img": [
        "fiore",
        "fiore",

    ],
    "Eagle_img": [
        "acquila",
        "aquila",

    ],
    "Edelweiss_img": [
        "stella alpina",


        "stella alpina in ambito naturale",
    ],
    "german_shepherd_img": [
        "cane",
        "cane",

    ],
    "orange_tree_img": [
        "arancio",
        "arancio",

    ],
    "Ostrich_img": [
        "struzzo",
        "struzzo",
    ],
    "owl_img": [
        "gufo",
        "gufo",

        "gufo appolaiato",
    ],
    "penguin_img": [
        "pinguino",
        "pinguino",
        "pinguino",
    ],
    "persian_img": [
        "gatto",
        "gatto",

    ],
    "Pine_img": [

    ],
    "rose_img": [
        "rosa",
        "rosa",

    ],
    "siamese_img": [

    ],
    "tulip_img": [

    ],
    "cerberus_img": [

        "cerbero",

        "schizzo preparatorio di una scena con kerbero il cane a tre teste che ha nelle zampe delle vittime",
    ],
    "cherubim_img": [

    ],
    "drow_img": [

    ],
    "forest_elf_img": [

    ],
    "golem_img": [

    ],
    "harpy_img": [
        "creatura mitologica",
        "creatura mitologica",

    ],
    "hellhound_img": [

    ],
    "hydra_img": [

        "idra",
        "Idra",

    ],
    "mermaid_img": [

    ],
    "pegasus_img": [
        "cavallo con ali",
        "cavallo alato",

        "cavallo alato disegnato",
    ],
    "phoenix_img": [

        "fenice",

        "araba fenice",
    ],
    "thunderbird_img": [

        "aquila",
        "Aquila",

    ],
    "triton_img": [

        "tritone",

        "tritone",
    ],
    "troll_img": [

        "troll",

        "troll",
    ],
    "unicorn_img": [
        "unicorno",
        "unicorno",

    ],
    "wyvern_img": [

    ],
    "Brocolli_img": [
        "Broccolo",
        "Broccole",
        "broccoli",

        "broccoli",
    ],
    "Carrot_img": [
        "Carote",
        "Carota",
        "carote",

        "carote",
    ],
    "Cauliflower_img": [
        "Cavolfiore",
        "cavolfiore",
        "cavolfiore",
        "un cavolfiore",
        "Coliflor",
        "cavolfiore",
    ],
    "Chicken_Breast_img": [

        "Pollo",
        "pollo",

    ],
    "Apple_img": [

        "Mela",
        "mela",

    ],
    "Ham_slice_img": [
        "Prosciutto affumicato",
        "Prosciutto",

    ],
    "McIntosh_Apple_img": [

        "Mela",
        "mela",
        "una mela",

    ],
    "moon_drop_grape_img": [

        "Uva",
        "uva",

        "uva",
    ],
    "pork_sausage_img": [

        "Salsiccia",
        "salsiccia",
        "delle salsicce",

    ],
    "red_globe_grape_img": [

        "Uva",
        "uva",

        "uva",
    ],
    "ribeye_steak_img": [

        "bistecca",
        "una bistecca di manzo",

    ],
    "salmon_fillet_img": [

        "Salmone",
        "salmone",

        "salmone",
    ],
    "sweet_potato_img": [
        "Patate",

        "patate",

    ],
    "t_bone_steak_img": [

        "Bistecca",

        "una bistecca di manzo",
    ],
    "tuna_steak_img": [
        "trancio tonno",

        "un trancio di tonno",

    ],
    "turkey_breast_img": [


        "arrosto",

        "arrosto",
    ],
    "bass_trombone_img": [
        "trombone",


        "trombone",
    ],
    "bell_img": [
        "Campanella",
        "Campana",
        "campanella",
        "una campana",
        "campana",
    ],
    "bongo_drums_img": [
        "Bonghi",
        "Bongos",
        "bonghi",

    ],
    "castanets_img": [
        "Nacchere",

        "nacchere",

        "CastaÃ±uela",
        "nacchere",
    ],
    "clarinet_img": [
        "Clarinetto",
        "Clarinetto",
        "clarinetto",
        "un clarinetto",
        "clarinetto",
    ],
    "cornet_img": [

        "Tromba",
        "tromba",
        "una tromba",
        "tromba",
    ],
    "fife_img": [
        "Flauto",
        "Flauto",
        "flauto",

    ],
    "piccolo_img": [

        "clarinetto",

        "clarinetto",
    ],
    "roto_drums_img": [

        "batteria",
        "una batteria",
        "batteria",
    ],
    "saxophone_img": [

        "Sassofono",
        "sassofono",
        "un sassofono",
        "saxofono",
    ],
    "slit_drum_img": [
        "Marimba",
        "Marimba",

    ],
    "tambourine_img": [
        "Tamburello",
        "Un tamburello",
        "tamburello",
        "un tamburello",

        "tamburello",
    ],
    "tenor_trombone_img": [
        "Trombone pistoni",
        "Trombone",

        "trombone",
    ],
    "timpani_img": [

        "Tambore",
        "tamburo",
        "un tamburo",

    ],
    "triangle_img": [
        "Triangolo",
        "Triangolo",
        "triangolo",
        "un triangolo",
        "triangolo",
    ],
    "trumpet_img": [
        "Tromba",
        "Tromba",
        "tromba",
        "una tromba",
        "tromba",
    ],
    "3D_printer_img": [

        "stampante 3d",
        "Stampante 3d",
        "stampante 3D",
    ],
    "alienware_desktop_img": [
        "altoparlante",

        "Altoparlante",

    ],
    "alienware_laptop_img": [

        "computer portatile",

        "computer portatile",
    ],
    "CRT_monitor_img": [

        "monitor",
        "Monitor",
        "monitor",
    ],
    "iMac_img": [

        "monitor",
        "Monitor",
        "monitor",
    ],
    "LCD_monitor_img": [

        "monitor",
        "Monitor",
        "monitor",
    ],
    "mac_pro_img": [

    ],
    "macbook_air_img": [

        "computer portatile",

        "computer portatile",
    ],
    "macbook_pro_img": [

        "computer portatile",

        "computer portatile",
    ],
    "wired_keyboard_img": [
        "tastiera",
        "tastiera",
        "Tastiera",
        "tastiera",
    ],
    "mouse_img": [
        "mouse",
        "mouse",
        "Mouse",
        "mouse",
    ],
    "wireless_keyboard_img": [
        "tastiera",
        "tastiera",
        "Tastiera",
        "tastiera",
    ],
    "wireless_mouse_img": [
        "mouse",
        "mouse",
        "Mouse",
        "mouse",
    ],
    "XPS_desktop_img": [

        "computer",
        "Computer",

    ],
    "XPS_laptop_img": [
        "computer",

        "Computer",

    ],
    "chainsaw_img": [

        "motosega",

        "motosega",
    ],
    "chisel_img": [

    ],
    "drain_shovel_img": [

        "pala",
        "Pala",
        "pala",
    ],
    "electric_pruner_img": [
        "cesoie",
        "cesoia",
        "Cesoia",

    ],
    "flat_brush_img": [
        "pennello",
        "pennello",
        "Pennello",
        "pennello",
    ],
    "gardening_shear_img": [
        "cesoie",
        "cesoia",
        "Cesoia",

    ],
    "glass_palette_img": [
        "tavolozza",
        "tavolozza",

    ],
    "lopper_img": [
        "cesoie",
        "cesoia",
        "Cesoie",

    ],
    "mallet_img": [

    ],
    "potting_trowel_img": [
        "pala da giardinaggio",

        "paletta da giardinaggio",
    ],
    "rake_img": [
        "rastrello",
        "rastrello",
        "Radtrello",
        "rastrello",
    ],
    "rockery_trowel_img": [

    ],
    "round_brush_img": [
        "pennello",
        "pennello",
        "Prnnello",
        "pennello",
    ],
    "rounded_point_shovel_img": [

        "pala",
        "Pala",
        "pala",
    ],
    "wire_img": [
        "cavo",

        "Cavo",

    ],
    "wood_palette_img": [
        "tavolozza",
        "tavolozza",
        "Tavolozza",

    ],
}













# =========================
# Normalization
# =========================
# Clean the dataframe labels
df['Image'] = df['Image'].str.strip().str.lower()

# Clean the correct_answers dictionary
correct_answers_normalized = {
    k.lower().strip(): [ans.lower().strip() for ans in v]
    for k, v in correct_answers.items()
}

# =========================
# Agreement Rate Calculation with Debugging
# =========================
image_agreement = {}
missing_images = []
unmatched_responses = {}

for image in df['Image'].unique():
    if image not in correct_answers_normalized:
        missing_images.append(image)
        continue

    image_df = df[df['Image'] == image]
    responses = image_df['Response'].dropna().str.strip().str.lower()

    acceptable_answers = correct_answers_normalized[image]

    agree_count = 0
    disagree_count = 0
    unmatched_resp = []

    for resp in responses:
        if resp in acceptable_answers:
            agree_count += 1
        else:
            disagree_count += 1
            unmatched_resp.append(resp)

    total = agree_count + disagree_count
    agreement_rate = (agree_count / total) * 100 if total > 0 else 0

    image_agreement[image] = {
        'Agreement %': agreement_rate,
        'Agree Count': agree_count,
        'Disagree Count': disagree_count,
        'Total Responses': total
    }

    if unmatched_resp:
        unmatched_responses[image] = Counter(unmatched_resp)

# =========================
# Chi-Square Test
# =========================
# Prepare contingency table
contingency_data = []
labels = []

for image, data in image_agreement.items():
    if data['Total Responses'] >= 5:  # Ensure enough sample size
        contingency_data.append([data['Agree Count'], data['Disagree Count']])
        labels.append(image)

# Run chi-square if we have enough images
if len(contingency_data) >= 2:
    chi2, p, dof, expected = chi2_contingency(contingency_data)

    print("\nChi-Square Test Results")
    print(f"Chi2 Statistic: {chi2}")
    print(f"p-value: {p}")
    print(f"Degrees of Freedom: {dof}")
    print(f"Expected Frequencies:\n{expected}")

    # Interpret
    alpha = 0.05
    if p < alpha:
        print("\nConclusion: There is a statistically significant difference in agreement rates across images (reject null hypothesis).")
    else:
        print("\nConclusion: No significant difference in agreement rates across images (fail to reject null hypothesis).")
else:
    print("\nNot enough images with sufficient responses to perform chi-square test.")

# =========================
# Outputs
# =========================
print("\nAgreement Rates per Image:")
for image, stats in image_agreement.items():
    print(f"{image}: {stats['Agreement %']:.2f}% (n={stats['Total Responses']})")

if missing_images:
    print("\nMissing Images from Dictionary:")
    print(missing_images)

if unmatched_responses:
    print("\nUnmatched Responses by Image:")
    for image, counts in unmatched_responses.items():
        print(f"{image}: {dict(counts)}")
```
