```python
import pandas as pd
from collections import Counter
from scipy.stats import chi2_contingency
import numpy as np

# Load your Excel file (replace the path with your actual file)
file_path = '/content/sample_data/nonexperts_PT_clean (2).xlsx'
df = pd.read_excel(file_path)

# Example correct answers dictionary
correct_answers = {
    "Brocolli_img": [

        "Brócoli",
        "Brócolis",
        "Brócolis",
        "Brócolis",
        "brocoli",
        "Brócolis",
        "Brócolis",
        "Brocolis",
        "Brócolis",
        "Brócolis",
        "Brócoli",
        "Brócolis",
        "brocolis",
        "Brócolis",

        "Brócolis",
        "brocolis",
        "brócolis",
    ],
    "Carrot_img": [

        "Cenoura",
        "Cenoura",
        "Cenoura",
        "Cenoura",
        "zanahoria",
        "Cenoura",

        "Cenoura",
        "Cenoura",
        "Cenoura",
        "Cenoura",
        "Cenoura",
        "cenoura",
        "Cenoura",

        "Cenouras",
        "Cenoura",

        "Cenouras",
    ],
    "Cauliflower_img": [

        "Couve flor",
        "Couve-flor",
        "Couve-flor",
        "Couve flor",
        "coliflor",
        "Couve flor",
        "Couve flor",
        "Couve-flor",
        "Couve flor",
        "Couve-flor",
        "Couve flor",
        "Couve flor",
        "couve-flor",
        "Couve-flor",
        "couve flor",
        "Couve-flor",
        "Couve- flor",
        "couve-flor",
        "couve flor",
        "Couve-flor",
    ],
    "Chicken_Breast_img": [

        "Peito de frango",

        "Peito de frango",

        "Peito de frango",
        "Peito de frango",
        "Peito de frango",
        "Peito de frango",


        "Peito de frango",

        "Peito de frango",

        "Peito de frango",

        "peite de frango",
    ],
    "Apple_img": [

        "maça verde",
        "Maça verde",
        "maça verde",
        "Maçã  verde",
        "manzana verde",
        "Maçã verde",
        "Maçã verde",
        "Maçã verde",
        "Maçã verde",
        "Maçã verde",

        "Maçã verde",

        "Maçã verde",
        "Maçã verde",

        "Maçã verde",
        "maçã verde",

    ],
    "Ham_slice_img": [

        "Presunto",

        "Presunto",

        "Presunto",
        "Presunto",
        "Presunto",


        "Presunto",
        "Presunto",

        "Presunto",

    ],
    "McIntosh_Apple_img": [

        "Maça",
        "Maça",
        "maça",
        "Maçã",
        "manzana",
        "Maçã",
        "Maçã",
        "Maça",
        "Maçã",
        "Maçã",
        "Maçã",
        "Maçã",
        "maçã",
        "Maçã",
        "Uma maçã",
        "Maça",
        "maçã",
    ],
    "moon_drop_grape_img": [

        "Uva",

        "Uva",


        "Uvas",
        "Uva",
        "Uvas",
        "Uva",

        "uva",

        "Uvas",
        "Uva",

    ],
    "pork_sausage_img": [

        "Linguiça",
        "Linguiça",

        "Linguica",

        "Linguiça",
        "Linguiça",
        "Linguiça",
        "Linguiça",
        "Linguiça",
        "linguiça",

        "Linguiça",
        "Linguiça",
        "Linguiça",
        "linguiça",
        "linguiça",
    ],
    "red_globe_grape_img": [

        "Uva",

        "uva",
        "Uva",
        "uvas",
        "Uva",

        "Uva",
        "Uva",
        "Uvas",
        "Uva",

        "uva",

        "Uvas",
        "Uvas",
        "uva",

    ],
    "ribeye_steak_img": [


        "Carne",
        "Carne",
        "Carne",

        "Carne",

    ],
    "salmon_fillet_img": [

        "Salmão",
        "Salmão",
        "Salmão",
        "salmón",
        "Salmao",
        "Salmão",

        "Salmão",
        "Salmão",
        "Salmão",

        "Salmão",

        "Salmão",
        "salmão",
    ],
    "sweet_potato_img": [

        "Batata doce",
        "Batata doce",
        "Batata doce",
        "Batata doce",

        "Batata doce",
        "Batata doce",
        "Batata doce",
        "Batata doce",
        "Batata doce",

        "batata doce",
        "Batata doce",
        "Batata doce",
        "Batata doce",
        "Batata doce",
        "batata doce",
        "Batata doce",
    ],
    "t_bone_steak_img": [


        "t-bone",
        "T bone",

        "t bone",

        "t-bone steak",
        "T boné carne",

    ],
    "tuna_steak_img": [


        "atún",
        "Atum",

        "Atum",
        "Atum",
        "Atum",

        "atum",
    ],
    "turkey_breast_img": [

        "Peito de peru",

        "Peito de peru",


        "peito de peru",

    ],
    "bass_trombone_img": [

        "Trombone",
        "Trombone",



        "Trombone",

        "Trombone",
        "trombone",

        "Trombone",
        "Trombone",
        "Trombone",

        "trombone",
    ],
    "bell_img": [

        "Sino",
        "Sino",
        "sino",
        "Sino",

        "Sino",
        "Sino",
        "Sino",
        "Sino",
        "Sino",
        "Sino",
        "Sino",

        "Sino",
        "Sino",
        "Sino",
        "Sino",
        "sino",
        "Sino",
    ],
    "bongo_drums_img": [


        "Bongô",

        "Bongô",
        "bongos",

        "Bongo",

        "Bongo",
        "bongô",

        "Bongô",

    ],
    "castanets_img": [

        "Castanhola",

        "castanholas",
        "Castanholas",
        "castañuelas",
        "Castanhola",
        "Castanholas",

        "Castanhola",
        "Castanholas",
        "Castanholas",
        "Castanholas",
        "castanholas",
        "Castanhola",
        "Castanholas",

        "Castanhola",

        "Castanholas",
    ],
    "clarinet_img": [


        "Clarinete",


        "clarinete",

        "Clarinete",
        "Clarinete",
        "Clarinete",
        "Clarinete",

        "Claronete",

        "Clarinete",
        "Clarinete",
        "clarinete",
    ],
    "cornet_img": [

        "Trompete",

        "Trompete",

        "Trompete",
        "Trompete",
        "Trompete",

        "Trompete",
        "trompete",
        "Trompete",
        "Trompete",

        "Trompete",
        "trompete",

    ],
    "fife_img": [

        "Flauta",
        "Flauta",


        "flauta",
        "Flauta",


        "Flauta",
        "Flauta",
        "Flauta",

        "Flauta",
        "Uma flauta",


        "flauta",

        "flauta",
        "Flauta",
    ],
    "piccolo_img": [

        "Flauta",

        "tipo flauta",

        "Flauta",

        "Flauta",

        "Flauta",

        "Flauta",

        "Flauta",

        "flauta",

    ],
    "roto_drums_img": [


        "Bateria",


        "Bateria",



        "Bateria",

        "Bateria",
        "bateria",

    ],
    "saxophone_img": [

        "Saxsophone",
        "Saxofone",
        "Saxofone",
        "Saxofone",
        "saxofón",
        "Saxofone",
        "Saxsofone",
        "Saxofone",
        "Saxofone",
        "Saxofone",
        "Sacofone",
        "Saxofone",
        "saxofone",
        "Saxofone",
        "Saxofone",
        "Saxofone",
        "Sax",
        "saxofone",
        "Saxofone",
        "Saxofone",
    ],
    "slit_drum_img": [


        "Instrumento",



        "Instrumento musical",
        "Instrumento musical",


    ],
    "tambourine_img": [

        "pandeiro",
        "Pandeiro",

        "Pandeiro",

        "Pandeiro",
        "Pandeiro",
        "Pandeiro",
        "Pandeiro",
        "Pandeiro",
        "Pandeiro",
        "Pandeiro",
        "pandeiro",
        "Pandeiro",
        "Um pandeiro",
        "Pandeiro",
        "Pandeiro",
        "pandeiro",

    ],
    "tenor_trombone_img": [


        "Trombone",

        "Trombone",


        "Trombone",

        "Trombone",

        "Trombete",
        "Trombeta",
        "Trombone",
        "Trombone",
        "trombone",
    ],
    "timpani_img": [

        "Tambor",
        "Tambor",


        "Tambor",

        "Tambor",

        "Tambor",

        "Um tambor",
        "Tamborim",


    ],
    "triangle_img": [

        "Triangulo",
        "Triângulo",

        "Triângulo",
        "triángulo",
        "Triangulo",

        "triângulo",
        "Triângulo",
        "Triângulo",

        "Triângulo",
        "triangulo (musica)",

        "Triângulo",
        "Triângulo",
        "Triángulo",
        "triangulo",
        "Triangulo",
        "Triângulo",
    ],
    "trumpet_img": [


        "Trompete",

        "trompeta",
        "Trompete",
        "Um trompete",

        "Trompete",
        "Trompete",

        "Trompete",
        "trompete",

        "Trompete",
        "Trompete",
        "Trompete",
        "trompete",
        "Trompete",
    ],
    "apple_tree_img": [
        "macieira",
        "Macieira",

        "macieira",

        "Macieira",

        "Macieira",
        "Macieira",
        "Macieira",

    ],
    "bulldog_img": [

        "Buldog",
        "Bulldog inglês",
        "bulldog",
        "Bulldog",

        "Bulldog",

        "Buldogue",
        "Cachorro Bulldog",
        "Bulldog",

        "Budog",

    ],
    "cedar_img": [
        "árvore",

        "uma árvore",

        "Árvore",

        "Arvore",
        "árvore",
        "Árvore",

        "Árvore",

        "Arvore",


        "Árvore",
        "Isto é uma Árvore",

    ],
    "dandelion_img": [
        "flor",

        "Flor",
        "flor",
        "Flor",
        "flor",
        "Flôr",
        "Flor",
        "Flor",

        "Flor",
        "Flor",

        "Flor",
        "Flor",
        "Flores",
    ],
    "Eagle_img": [

        "Uma águia",


        "Visão e águia",
        "Águia",
        "águia",
        "Aguia",
        "águia",
        "Águia",
        "Águia",
        "Aguia",

        "Aguia",
        "Águia",

        "Águia",
        "Águia",
        "Aguia",
    ],
    "Edelweiss_img": [
        "flor",

        "Flor",

        "Flor",
        "flor",
        "Flor",

        "Flor",
        "Flôr",
        "Flor",

        "Flor",

        "Flor",

    ],
    "german_shepherd_img": [

        "Pastor alemão",
        "um pastor alemão",


        "Pastor alemão",
        "pastor alemao",

        "Pastor alemão",

        "Pastor-alemão",

        "Pastor Alemão",

        "Pastor alemão",

    ],
    "orange_tree_img": [

        "Uma laranjeira",
        "Laranjeira",


        "Laranjeira",
        "laranjeira",
        "Laranjeira",
        "Laranjeira",
        "Laranjeira",
        "Laranjeira",
        "Laranjeira",
        "Laranjeira",
        "Uma laranjeira",
        "Laranjal",

        "Fartura",
    ],
    "Ostrich_img": [

        "Avestruz",
        "Avestruz",
        "avestruz",
        "Avestruz",
        "Avestruz",
        "avestruz",
        "Avestruz",
        "avestruz",
        "Avestruz",
        "Avestruz",
        "Avestruz",
        "Avestruz",

        "Avestruz",
        "Um avestruz",
        "Avestruz",
        "Avestrus",
    ],
    "owl_img": [
        "coruja",
        "Coruja",
        "Coruja",

        "Coruja",
        "coruja",
        "Coruja",
        "coruja",
        "Coruja",
        "Coruja",
        "Coruja",
        "Coruja",

        "Coruja",

        "Coruja",
        "Coruja",
        "Coruja",
        "Coruja",
    ],
    "penguin_img": [
        "pinguim",
        "Pinguim",
        "Pinguim",
        "pinguim",
        "Pinguim",
        "Pinguim",
        "pinguim",
        "Pinguïn",
        "pinguin",
        "Pinguim",
        "Pinguim",
        "Pinguim",
        "Pinguim",
        "Pinguim",
        "Pinguim",

        "Pinguim",
        "Pinguim",
    ],
    "persian_img": [
        "gato",


        "Gato",
        "gato",
        "Gato",


        "Gato",

        "Gato",

        "Gato",

    ],
    "Pine_img": [
        "árvore",

        "a´rvore",

        "Árvore",

        "Arvore",

        "arvore"



    ],
    "rose_img": [
        "rosa",

        "Rosa",
        "uma rosa",

        "Rosa",
        "rosa",
        "Rosa",

        "Rosa",
        "Rosa",
        "Rosa",

        "rosa",
        "Rosa",

    ],
    "siamese_img": [
        "gato",

        "Gato",

        "Gato",
        "gato",
        "Gato",
        "gato",
        "Gato",
        "Gato",
        "Gato",

        "gato",

        "Gata",
        "Gato",
        "Gato",
    ],
    "tulip_img": [

        "Flor",

        "flor",

        "Flôr",
        "Flor",
        "Flor",

        "Flor",

        "Flor",
        "Uma flor.",
    ],
    "cerberus_img": [

        "Cérbero",
        "Cerberus",
        "cérbero",


        "Cerberus",

        "Cerberus",
        "Cérbero",

    ],
    "cherubim_img": [

        "Anjo",
        "Anjo",
        "um anjo, possivelmente um querubim",

        "Anjo",
        "anjo",
        "Anjo querubim",
        "anjo",
        "Anjo",

        "Anjo",

        "Anjo",

        "Anjo",
        "Um anjo",

        "Anjo",
    ],
    "drow_img": [

        "Elfo",

        "elfo",
        "Elfo",
        "elfa",

        "Elfa",
        "Elfa",

        "Élfo",
    ],
    "forest_elf_img": [

        "Arqueira",
        "Elfo arqueiro",

        "Elfo arqueira",
        "Arqueiro",
        "Arqueira elfa",

        "Arqueira",

        "Arqueira",

    ],
    "golem_img": [

        "Coisa",

        "Coisa",

        "O Coisa",
        "coisa",

        "Coisa",

        "O Coisa",

    ],
    "harpy_img": [

        "uma harpia",

        "Arpia",

        "Harpia",
        "Harpia",
        "Harpia",

        "harpia",

    ],
    "hellhound_img": [

        "Cachorro cuspindo fogo",


        "Cachorro cuspindo fogo",

        "Cachorro cuspindo fogo",

        "Cachorro, cuspindo fogo",
    ],
    "hydra_img": [
        "dragão de várias cabeças",

        "Dragão de várias cabeças",

        "Dragão de 7 cabeças",

        "Dragão de 7 cabeças",
        "Dragão de 7 cabeças",

        "Um dragão com sete cabeças",

        "Dragão de 7 cabeças",

        "Dragão com 7 cabeças",
    ],
    "mermaid_img": [
        "sereia",
        "Uma sereia",
        "Sereia",
        "uma sereia",

        "Sereia",
        "sereia",
        "Sereia",
        "sereia",
        "Sereia",
        "Sereia",
        "Sereia",

        "sereia",
        "Sereia",
        "Uma sereia",
        "Sereia",
        "Sereia",
        "Sereia",
    ],
    "pegasus_img": [
        "pégasus",
        "Pegasus",
        "Pegasus",
        "pegasu",


        "pegaso",
        "Pegasus",
        "pegasus",
        "Pegasus",
        "Pegasus",
        "Pegasus",

        "Pegasus",
        "Pegasus",

        "Pégasus",

    ],
    "phoenix_img": [

        "Uma fênix",
        "Fênix",
        "ave fênix",

        "Fênix",

        "Phoenix",
        "fênix",
        "Fênix",
        "Fênix",

        "Fenix",
        "Fenix",

        "Uma fênix",

        "Fenix",
        "Fenix",
    ],
    "thunderbird_img": [
        "águia",

        "Águia",
        "uma águia",

        "Águia",

        "Aguia",

        "Águia",
        "Águia",

        "Águia",
        "Uma águia",

        "Águia",
        "Aguia",
    ],
    "triton_img": [
        "Poseidon",
        "Poseidon",
        "Poseidon",
        "poseidon",

        "Poseidon",



        "Poseidom",

    ],
    "troll_img": [
        "ogro",

        "ogro",

        "Ogro",


        "Ogro",

        "Um ogro",
        "Ogro",

    ],
    "unicorn_img": [
        "unicórnio",
        "Um unicórnio",
        "Unicórnio",
        "unicórnio",

        "Unicórnio",
        "unicórnio",
        "Unicórnio",
        "unicornio",
        "Unicórnio",

        "Unicórnio",
        "Unicórnio",
        "Unicornio",
        "Unicórnio",
        "Um unicórnio",
        "Unicórnio",
        "Unicornio",

        "Unicórnio",
    ],
    "wyvern_img": [
        "dragão",

        "Dragão",
        "dragão",
        "Dragão",
        "dragão",
        "Dragão",


        "dragão",
        "Dragão",

        "Dragão",

    ],
    "3D_printer_img": [

        "Impressora 3D",
        "Impressora 3D",

        "impressora 3D",
        "Impressora 3d",
        "Impressora 3d",
        "3D printer",

        "Impressora 3d",

        "Impressão 3d",

        "Impressora 3D",
        "Impressora 3D?",

    ],
    "alienware_desktop_img": [

        "Caixa de som",
        "Caixa de som",
        "Caixa de som",

        "caixa de som",
        "Caixa de som",
        "Caixa de som",

        "Caixa de som",

    ],
    "alienware_laptop_img": [

        "Notebook",


        "Notebook",

        "Notebook",
        "Notebook",

        "Notebook",

        "Notebook",
    ],
    "CRT_monitor_img": [

        "Monitor",
        "Monitor",
        "Monitor",
        "Monitor",
        "monitor",
        "Monitor",

        "Monitor",

        "Monitor",
        "Monitor",

        "Monitor",

    ],
    "iMac_img": [
        "Monitor",

        "Monitor",
        "Monitor",
        "Monitor",
        "monitor",
        "Monitor",

        "Monitor",

        "Monitor",

        "Monitor",
    ],
    "LCD_monitor_img": [
        "Monitor",
        "Monitor",
        "Monitor",
        "Monitor",
        "Monitor",
        "monitor",
        "Monitor",

        "Monitor",
        "Monitor",
        "Monitor",
        "Monitor",
        "monitor",
        "Monitor",

        "Monitor",

        "Monitor",
    ],
    "mac_pro_img": [

        "Mac PC",

        "Macintosh",

    ],
    "macbook_air_img": [
        "Notebook",

        "Notebook",
        "Notebook",
        "Notebook",

        "Notebook",

        "Notebook",
        "Notebook",

        "notebook",
        "Notebook",

        "Notebook",

        "Notebook",
    ],
    "macbook_pro_img": [
        "Notebook",

        "Notebook",
        "Notebook",
        "Notebook",

        "Notebook",

        "Notebook",
        "Notebook",

        "notebook",
        "Notebook",

        "Notebook",

        "Notebook",
    ],
    "wired_keyboard_img": [
        "Teclado",
        "Teclado",

        "Teclado",
        "Teclado",
        "teclado",
        "Teclado",

        "Teclado",
        "Teclado",
        "Teclado",
        "Teclado",
        "teclado",
        "Teclado",

        "Teclado",
        "Teclado",

    ],
    "mouse_img": [
        "Mouse",
        "Mouse",
        "Mouse",
        "Mouse",
        "Mouse",
        "Mouse",
        "Mouse",


        "Mouse",
        "Mouse",
        "Um mouse",
        "mouse",
        "Mouse",

        "Mouse",

    ],
    "wireless_keyboard_img": [
        "Teclado",

        "teclado",
        "Teclado",

        "Teclado",

        "Teclado",
        "Teclado",
        "teclado",
        "Teclado",

        "Teclado",
        "Teclado",
        "Teclado",
    ],
    "wireless_mouse_img": [
        "Mouse",
        "Moise",

        "Mouse",

        "mouse",
        "Mouse",
        "Mouse",


        "Mouse",
        "Um mouse",
        "mouse",
        "Mouse",

        "Mouse",
        "Um mouse",
        "Mouse",
    ],
    "XPS_desktop_img": [

        "CPU",

        "Cpu",
        "Cpu",

        "cpu",

        "Cpu de computador",
        "Cpu",

    ],
    "XPS_laptop_img": [
        "Notebook",

        "Notebook",
        "Notebook",
        "Notebook",

        "Notebook",
        "Notebook",

        "Notebook",
        "Notebook",

        "notebook",
        "Notebook",

        "Notebook",
    ],
    "chainsaw_img": [
        "Motosserra",
        "Motoserra",
        "Motoserra",

        "Motoserra",

        "Motoserra",

        "Moto serra",

        "Motosserra",
        "Motosserra",

    ],
    "chisel_img": [
        "Talhadeira",
        "Talhadeira",

        "Talha",

        "Talhador",

    ],
    "drain_shovel_img": [
        "Pá",

        "Pá",
        "Pá",
        "Pá",
        "pá",
        "Pá",

        "Pá",

        "Pá",
        "Uma pá",

        "Pá",
        "Pa",

        "Pá",
    ],
    "electric_pruner_img": [
        "Alicate",




        "Alicate",
        "Alicate",

    ],
    "flat_brush_img": [
        "Pincel",
        "Pincel",
        "Pincel",
        "Pincel",
        "Pincel",
        "pincel",
        "Pincel",
        "Pincel",
        "Pincel",

        "Pincel",
        "Pincel",
        "pincel",
        "Pincel",

        "Pincel",

    ],
    "gardening_shear_img": [


        "Alicate",


        "Alicate",

        "Alicate",

    ],
    "glass_palette_img": [

        "tintas",

        "Tábua tinta",

        "Tintas usadas",

        "tinta",
        "Tintas"



        "Tinta",

    ],
    "lopper_img": [

        "Alicate",

        "Alicate",
        "Alicate",
        "alicate cortador",
        "Alicate",
        "Alicate grande",

        "Alicate de corte",
        "Outro tipo de slicate",
        "Alicate",
    ],
    "mallet_img": [
        "Martelo de carne",


        "Martelo de carne",

        "martelo de carne",
        "Martelo de carne",

        "Martelo para carne",
        "Martelo de carne",


        "martelo de bater carne",

        "Martelo de carne",
        "Martelinho de carne",
        "Martelo para carne",
    ],
    "potting_trowel_img": [


        "Pá",
        "Pá",

        "Pá",
        "Pá",
        "Pá",
        "pá",

        "Pa",

    ],
    "rake_img": [
        "Restelo",

        "Rastelo",
        "Rastelo",
        "rastelo",
        "Rastelo",
        "Rastelo",
        "Rastelo",

        "Rastel",

        "Rastel",
        "Rastilho",
        "Rastel",
    ],
    "rockery_trowel_img": [

        "Ferramenta de jardinagem",



        "Ferramenta",

        "Ferramenta para jardinagem",
        "Ferramenta",

    ],
    "round_brush_img": [
        "Pincel",
        "Pincel",
        "Pincel",
        "Pincel",
        "Pincel",
        "pincel",
        "Pincel",
        "Pincel",
        "Pincel",
        "Pincel",
        "Pincel",
        "Pincel",
        "pincel",
        "Pincel",
        "Pincel",
        "Pincel",
        "Pincel",
        "Pincel",

    ],
    "rounded_point_shovel_img": [
        "Pá",
        "Pá",
        "Pá",
        "Pá",
        "Pá",
        "pá",
        "Pá",
        "Pá",
        "Pá",

        "Pá",
        "Pá",
        "pá",
        "Pá",
        "Pa",

        "Pá",
        "Pa",

    ],
    "wire_img": [

        "Arame",

        "arame",
        "Arame",

        "Arame",
        "Arame",
        "Arame",
        "Arame",
        "arame",
        "Arame",
        "Arame",

        "Arame",

    ],
    "wood_palette_img": [



        "Paleta de pintura",

        "Paleta de pintura",

        "Paleta de madeira",


        "Paleta de pintura",

    ],
}













# =========================
# Normalization
# =========================
# Clean the dataframe labels
df['Image'] = df['Image'].str.strip().str.lower()

# Clean the correct_answers dictionary
correct_answers_normalized = {
    k.lower().strip(): [ans.lower().strip() for ans in v]
    for k, v in correct_answers.items()
}

# =========================
# Agreement Rate Calculation with Debugging
# =========================
image_agreement = {}
missing_images = []
unmatched_responses = {}

for image in df['Image'].unique():
    if image not in correct_answers_normalized:
        missing_images.append(image)
        continue

    image_df = df[df['Image'] == image]
    responses = image_df['Response'].dropna().str.strip().str.lower()

    acceptable_answers = correct_answers_normalized[image]

    agree_count = 0
    disagree_count = 0
    unmatched_resp = []

    for resp in responses:
        if resp in acceptable_answers:
            agree_count += 1
        else:
            disagree_count += 1
            unmatched_resp.append(resp)

    total = agree_count + disagree_count
    agreement_rate = (agree_count / total) * 100 if total > 0 else 0

    image_agreement[image] = {
        'Agreement %': agreement_rate,
        'Agree Count': agree_count,
        'Disagree Count': disagree_count,
        'Total Responses': total
    }

    if unmatched_resp:
        unmatched_responses[image] = Counter(unmatched_resp)

# =========================
# Chi-Square Test
# =========================
# Prepare contingency table
contingency_data = []
labels = []

for image, data in image_agreement.items():
    if data['Total Responses'] >= 5:  # Ensure enough sample size
        contingency_data.append([data['Agree Count'], data['Disagree Count']])
        labels.append(image)

# Run chi-square if we have enough images
if len(contingency_data) >= 2:
    chi2, p, dof, expected = chi2_contingency(contingency_data)

    print("\nChi-Square Test Results")
    print(f"Chi2 Statistic: {chi2}")
    print(f"p-value: {p}")
    print(f"Degrees of Freedom: {dof}")
    print(f"Expected Frequencies:\n{expected}")

    # Interpret
    alpha = 0.05
    if p < alpha:
        print("\nConclusion: There is a statistically significant difference in agreement rates across images (reject null hypothesis).")
    else:
        print("\nConclusion: No significant difference in agreement rates across images (fail to reject null hypothesis).")
else:
    print("\nNot enough images with sufficient responses to perform chi-square test.")

# =========================
# Outputs
# =========================
print("\nAgreement Rates per Image:")
for image, stats in image_agreement.items():
    print(f"{image}: {stats['Agreement %']:.2f}% (n={stats['Total Responses']})")

if missing_images:
    print("\nMissing Images from Dictionary:")
    print(missing_images)

if unmatched_responses:
    print("\nUnmatched Responses by Image:")
    for image, counts in unmatched_responses.items():
        print(f"{image}: {dict(counts)}")
```
